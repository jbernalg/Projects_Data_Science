# Mtodos de Seleccin de Features

La seleccin de features es un paso fundamental en el preprocesamiento de datos para modelos de Machine Learning. Consiste en identificar y seleccionar un subconjunto relevante de variables para usar en la construccin de modelos predictivos. La correcta seleccin de features mejora la precisin del modelo, reduce el sobreajuste, acelera el tiempo de entrenamiento y facilita la interpretacin del modelo.

En lo que sigue te presento los llamados **Mtodos de Wrapper**, los cuales evalan la calidad de los features utilizando un modelo de Machine Learning especfico. Estos suelen ser ms precisos, pero tambin ms costosos computacionalmente:

- **Eliminacin Recursiva de Features (RFE)**
- **Eliminacin Recursiva de Features con Validacin Cruzada (RFECV)**
- **Seleccin Secuencial de Features (SFS)**

Veamos de qu se trata cada uno

# 1.- Eliminacin Recursiva de Features (RFE)

Es una tcnica utilizada para seleccionar las variables ms importantes para un modelo. Su objetivo es eliminar las caractersticas menos relevantes, mejorando as el rendimiento del modelo y reduciendo su complejidad.

Generalmente, *RFE* se aplica a modelos que pueden proporcionar alguna medida de la importancia de las caractersticas como los modelos lineales o arboles de decisin. Entre los modelos donde se puede aplicar tenemos:

- Regresin Lineal.
- Regresin Logstica.
- Arboles de decisin.
- Bosques aleatorios.
- Maquina de vectores de soporte
- Entre otros...

## Cmo funciona RFE?

   Los pasos para entender el funcionamiento del mtodo *RFE* son los siguientes:
   
   1. **Entrenamiento inicial**: Entrena un modelo utilizando todas las caractersticas (Variables predictoras).
    
    
   2. **Evaluacin de importancia**: Evala la importancia de cada caracterstica medida de diferentes maneras dependiendo del modelo utilizado (coeficientes de la regresin lineal).
   
    
   3. **Eliminacin de caractersticas**: Elimina las caractersticas menos importantes para el modelo.
   
   
   4. **Repeticin de pasos**: Repite los pasos del 1 al 3 hasta alcanzar un nmero predefinido de caractersticas.

## Aplicacin de RFE

Vamos a aplicar *RFE* sobre un conjunto de datos de prueba de 6 variables, que incluye la variable objetivo, para encontrar las 3 ms importantes para un modelo de Rgresin Lineal.

```python
# Librerias para generar datos de prueba
import numpy as np
import pandas as pd

# semilla
np.random.seed(0)

# generamos los datos aleatorios de 5 caracteristicas
X = pd.DataFrame({
    'feature_1': np.random.rand(100),
    'feature_2': np.random.rand(100),
    'feature_3': np.random.rand(100),
    'feature_4': np.random.rand(100),
    'feature_5': np.random.rand(100)
})
```

Asociamos linealmente a la variable objetivo, 3 de las 5 caractersticas: feature 1, 2 y 5. Esto nos asegura que el modelo slo estar influenciado por estas tres caractersticas y entonces, al aplicar *RFE*, nos debe arrojar estas mismas en caso de funcionar correctamente.

```python
# variable objetivo
y = 3*X['feature_2'] + 2*X['feature_5'] + X['feature_1'] + np.random.rand(100)
```

Ahora, juntamos los datos en un Dataframe para mejorar la visualizacin.

```python
df = pd.DataFrame(X, columns=[f'feature_{i+1}' for i in range(X.shape[1])])
df['target'] = y
df.head()
```

### 1.- Entrenamiento inicial
Entrenamos el modelo de regresin lineal con todas las caractersticas.

```python
# libreria para usar la regresion lineal
from sklearn.linear_model import LinearRegression

# funcion para entrena el modelo
def train_model(X, y):
    model = LinearRegression()
    return model.fit(X,y)
```

```python
# entrenamos el modelo
modelo = train_model(X, y)
```

### 2.- Evaluacin de importancia
Para el caso de la regresin, debemos obtener los coeficientes de cada caractersticas una vez entrenado el modelo.

```python
# funcion para obtener los coeficientes de cada caracteristica
def evaluate_model(model):
    coeficientes = model.coef_
    return coeficientes
```

```python
# coeficiente de las caracteristicas
coefi = evaluate_model(modelo)
coefi
```

### 3.- Eliminacin de caractersticas
Identificamos la caracterstica con el coeficiente ms pequeo (el valor absoluto) y la eliminamos.

```python
# funcion que determina la caracteristica con menor valor de coeficiente y la elimina
def remove_feature(X, coeficientes):
    # indice del coficiente mas pequeo
    min_indice = np.argmin(np.abs(coeficientes))
    # nombre de la caracteristica correspondiente al indice anterior
    feature_remove = X.columns[min_indice]
    # eliminar caracteristica
    X = X.drop(columns=[feature_remove])
    # retornar matriz de caracteristica y nombre de la caracteristica eliminada
    return X, feature_remove
    
```

### 4.- Repetir hasta obtener el nmero deseado de caractersticas

Como queremos encontrar las 3 caracteristicas ms relvantes para el modelo, repetimos el proceso anterior tres veces. Para automatizar el proceso, procedemos como sigue:

```python
# numero de features a seleccionar
n_features = 3

# ejecutar proceso hasta llegar a 3 caracteristicas
while X.shape[1] > n_features:
    modelo = train_model(X,y)
    coefi = evaluate_model(modelo)
    X, feature_remove = remove_feature(X, coefi)
    
# mostrar las 3 caracteristicas mas importantes
print('Caracteristicas seleccionadas', X.columns)
    
    
```

> Los resultados confirman que los features 1, 2 y 5 son los mas importantea para el modelo de regresin lineal, tal como esperabamos. Y aqu redica la relevancia de este mtodo y es que nos permite encontrar, de un conjunto grande de caractersticas, un nmero especfico de ellas importantes para el modelo.

## Aplicacin de RFE con Scikit-Learn

Ahora vamos a utilizar la librera Scikit Learn para aplicar este mtodo y simplificar el proceso de eliminacin recursiva. Trabajarmos sobre los mismos datos de prueba del ejemplo anterior para hacer una comparacin de los resultados de ambos mtodos. 

```python
# features
X = df.drop(columns=['target'])
# variable objetivo
y = df['target']
```

Aplicamos RFE con la librera sklearn

```python
# libreria para aplicar RFE
from sklearn.feature_selection import RFE

# crear modelo de regresion
modelo_2 = LinearRegression()

# Crear el objeto RFE y especificar el numero de caracteristicas deseadas
rfe = RFE(estimator=modelo_2, n_features_to_select=3)

# ajustar RFE a los datos
rfe.fit(X, y)

# obtener las caracteristicas seleccionadas por RFE
features_select = X.columns[rfe.support_]

# mostrar nombre
features_select
```

> De esta forma, obtenemos las caractersticas ms importantes para el modelo de regresin, que en este caso son los features 1, 2 y 5, coincidiendo con los resultados del ejemplo anterior.

# 2.- Eliminacin Recursiva de Features con Validacin Cruzada (RFECV)

Es una tcnica de seleccin de caractersticas que mejora la metodologa *RFE* al incoporar la validacin cruzada para seleccionar automticamente el nmero ptimo de caractersticas.

## Cmo funciona RFECV?

1. **Entrenamiento del modelo**: Similar a RFE, entrena el modelo usando todas las caractersticas.


2. **Validacin cruzada**: Utiliza la validacin cruzada para evaluar el rendimiento del modelo en diferentes subconjuntos de datos.


3. **Eliminacin de caractersticas**: Elimina la caracterstica menos importante y evala el modelo usando validacin cruzada.


4. **Repeticin de pasos**: Repite los pasos del 1 hasta el 3 hasta que se encuentre el conjunto ptimo de caractersticas que maximiza el rendimiento del modelo.

## Aplicacin de RFECV con Scikit Learn

Vamos a utilizar un conjunto de datos de prueba de 11 caractersticas, incluyendo la variable objetivo, para aplicar RFECV y obtener las caractersticas mas importantes para un modelo de Regresin Lineal.  

```python
# semilla
np.random.seed(0)

# 10 caracteristicas
X = pd.DataFrame({
    'feature_1': np.random.rand(100),
    'feature_2': np.random.rand(100),
    'feature_3': np.random.rand(100),
    'feature_4': np.random.rand(100),
    'feature_5': np.random.rand(100),
    'feature_6': np.random.rand(100),
    'feature_7': np.random.rand(100),
    'feature_8': np.random.rand(100),
    'feature_9': np.random.rand(100),
    'feature_10': np.random.rand(100)
})
```

Relacionamos la variable objetivo con 4 caractersticas: feature 1, 2, 5 y 7. Esto asegura que de las 10 caracteristicas, estas 4 influencian el modelo y por tanto, al aplicar RFECV, nos debe arrojar las mismas.

```python
# variable objetivo
y = (
    3*X['feature_2'] + 
    2*X['feature_5'] + 
    X['feature_1'] + 
    X['feature_7']*X['feature_7'] +
    np.random.rand(100)
)
y.head()
```

Juntamos los datos en un Dataframe para visualizarlos mejor

```python
df = pd.DataFrame(X, columns=[f'feature_{i+1}' for i in range(X.shape[1])])
df['target'] = y
df.head()
```

Dividimos los datos en entrenamiento y prueba

```python
# libreria para dividir los datos
from sklearn.model_selection import train_test_split

# entrenamiento 70% y prueba 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

Aplicamos RFECV sobre el modelo de regresin lineal indicando el paso, el nmero de pliegues de la validacin cruzada y la mtrica a utilizar para medir el rendimiento del modelo.

```python
# importamos el metodo RFECV
from sklearn.feature_selection import RFECV

# creamos el modelo de regresion lineal
model_2 = LinearRegression()

# configuramos RFECV
rfecv = RFECV(
    estimator=model_2,
    step=1, # paso para eliminar caracteristicas
    cv=5, # numero de pliegues para la validacion cruzada
    scoring='neg_mean_squared_error' # uso del MSE negativo como metrica de rendimiento
)

# ajustar RFECV a los datos
rfecv.fit(X_train, y_train)

# mostrar numero optimo de caracteristicas
print('Numero optimo de caracteristicas: ', rfecv.n_features_)
# mostrar nombre de las caracteristicas
print('Caracteristicas seleccionadas: ', X_train.columns[rfecv.support_])
```

El RFECV arroja 4 variables ptimas para el modelo de Regresin Lineal: feature 1, 2, 5, y 7, mismas que han sido relacionadas al inicio con la variable objetivo y las esperadas como resultado.

RFECV recibe como parmetro el *scoring* que en este caso se asigna como el negativo del error cuadratico medio (MSE). Veamos una explicacin del porqu:
> *Esto se debe a como funciona la API y la convencin de algunas mtricas de evaluacin. El MSE es una mtrica de error, lo que significa que debemos minimizar este valor para mejorar el rendimiento del modelo. Sim embargo, la mayora de las funciones de Scikit learn estn diseadas para maximizar la mtrica de evaluacin. Por tanto, al usar MSE, Scikit Learn lo convierte en su valor negativo para que la funcin de maximizacin funcione correctamente. Al maximizar el negativo del MSE estamos minimizando el MSE real.*

# 3.- Seleccin Secuencial de Features (SFS)

Es una tcnica de seleccin de caractersticas usada para reducir el nmero de las mismas en un modelo de manera ms eficiente. Realiza la seleccin de manera secuencial, es decir, aade o elimina caractersticas una por una basandose en su impacto en el rendimiento del modelo.

## Cmo funciona SFS?

La implementacin de SFS va a depender del tipo de seleccin que se realice: Forward Selection (Seleccion hacia adelante) o Backward Selection (Seleccin hacia atrs). Vemos el paso a paso de cada mtodo:

### **Forward Selection**

1.- **Inicializacin**: Comienza con un modelo vaco, sin variables predictoras.

2.- **Evaluacin de caractersticas**: Evala todas las caractersticas individualmente y selecciona la que mejora el rendimiento del modelo.

3.- **Agregar caracerstica**: agrega la caracterstica seleccionada al modelo.

4.- **Repetir**: Repite el proceso, evaluando las caractersticas restantes en combinacin con las ya seleccionadas hasta alcanzar un nmero mximo de caractersticas determinadas.

### **Backward Selection**

1.-**Inicializacin**: Comienza con todas las caracteristicas disponibles.

2.- **Evaluacin de caractersticas**: Evala todas las caractersticas individualmente y selecciona la que tiene el menor impacto negativo en el rendimiento del modelo.

3.- **Eliminar caracterstica**: Elimina la caracterstica seleccionada del modelo.

4.- **Repetir**: Repite el proceso hasta alcanzar un nmero de caractersticas determinadas.

En general, la seleccin hacia adelante y hacia atrs no arroja resultados equivalentes debido a la forma en que van apareciendo las variables. Adems, una puede ser mucho ms rpida que la otra segn la cantidad solicitada de caractersticas: si tenemos 10 caractersticas y pedimos 7 caractersticas seleccionadas, la seleccin hacia adelante necesitara realizar 7 iteraciones mientras que la seleccin hacia atrs solo necesitara realizar 3.

## Diferencias entre SFS y RFE

- *SFS* ofrece tanto seleccin hacia adelante como hacia atrs, proporcinando mayor flexibilidad en la direccin del proceso de seleccin.

- *SFS* no requiere que el modelo exponga un atributo para mostrar las caractersticas seleccionadas.

- *SFS* puede ser ms eficiente en ciertos casos ya que evala incrementos especficos en el rendimiento del modelo.

- *RFE* puede ser computacionalmente costoso debido a la necesidad de recalcular la importancia de las caractersticas y reentrenar el modelo en cada iteracin.

- *SFS* puede ser ms lento ya que debe evaluar ms modelos.


## Aplicacin de SFS con Scikit Learn

Vamos a utilizar el conjunto de datos de prueba del ejemplo anterior y vamos aplicar el mtodo *SFS* tanto hacia adelante como hacia atrs.

```python
# semilla
np.random.seed(0)

# 10 caracteristicas
X = pd.DataFrame({
    'feature_1': np.random.rand(100),
    'feature_2': np.random.rand(100),
    'feature_3': np.random.rand(100),
    'feature_4': np.random.rand(100),
    'feature_5': np.random.rand(100),
    'feature_6': np.random.rand(100),
    'feature_7': np.random.rand(100),
    'feature_8': np.random.rand(100),
    'feature_9': np.random.rand(100),
    'feature_10': np.random.rand(100)
})
```

Relacionamos la variable objetivo con 4 caractersticas: feature 1, 2, 5 y 7. Esto asegura que de las 10 caracteristicas, solo 4 influencian el modelo y por tanto, al aplicar *SFS* slo algunas de estas, dependiendo de la cantidad que sean solicitadas, sern devueltas.

```python
# variable objetivo
y = (
    3*X['feature_2'] + 
    2*X['feature_5'] + 
    X['feature_1'] + 
    X['feature_7']*X['feature_7'] +
    np.random.rand(100)
)
y.head()
```

Dividimos los datos entrenamiento y prueba

```python
# libreria para dividir los datos
from sklearn.model_selection import train_test_split

# entrenamiento 70% y prueba 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

Aplicamos SFS hacia adelante para un modelo de regresin lineal y solicitamos las 4 caractersticas ms importantes.

```python
# libreria para aplicar SFS
from sklearn.feature_selection import SequentialFeatureSelector

# Crear el modelo de regresin lineal
model_3 = LinearRegression()

# configurar el selector de caracteristicas sencuencial
sfs = SequentialFeatureSelector(
    model_3,
    n_features_to_select=4, #cantidad de caracteristicas
    direction='forward', # direccion de seleccion
    cv=5 # numero de validaciones cruzadas
    )

# ajustamos el selector a los datos de entrenamiento
sfs.fit(X_train, y_train)

# obtener las caracteristicas seleccionadas
selected_features = X_train.columns[sfs.get_support()]

# mostras caracteristicas
selected_features
```

> Vemos que las caractersticas seleccionadas son las mismas que estn relacionadas con el modelo, por tanto, confirmamos que el mtodo funciona correctamente.

Ahora, aplicamos SFS hacia atrs para un modelo de regresin lineal y una cantidad de 3 caractersticas. Esta eleccin de parmetros, har que el procesos SFS demore un poco ms debido a que deben realizar ms iteraciones.

```python
# Crear el modelo de regresin lineal
model_4 = LinearRegression()

# configurar el selector de caracteristicas sencuencial
sfs = SequentialFeatureSelector(
    model_4,
    n_features_to_select=3, #cantidad de caracteristicas
    direction='backward', # direccion de seleccion
    cv=5 # numero de validaciones cruzadas
                               )

# ajustamos el selector a los datos de entrenamiento
sfs.fit(X_train, y_train)

# obtener las caracteristicas seleccionadas
selected_features = X_train.columns[sfs.get_support()]

# mostras caracteristicas
selected_features
```

> Las caractersticas seleccionadas por SFS hacia atrs estn relacionadas con la variable objetivo del modelo, por tanto, confirmamos que el mtodo funciona correctamente.
