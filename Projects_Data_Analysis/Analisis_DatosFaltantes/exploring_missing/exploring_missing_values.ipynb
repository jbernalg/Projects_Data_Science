#### 1. Tipo MCAR

Estos son datos faltantes ocasionados por herramientas que dejan de funcionar sin ninguna razon aparente. La localizacion de valores faltantes ocurre completamente al azar

```python
# ordena la variables de la qu mas tiene valores faltantes a la que menos tiene
(
    diabetes_df # dataframe
    .missing.sort_variables_by_missingness() # ordena las variable de mayor a menor cantidad de valores faltantes
    .pipe(missingno.matrix) # muestra el resultado como una matriz
)
```

Del grafico concluimos que en las variables bmi y glucose hay tan pocos valores faltantes que se pueden considerar valores faltantes al azar o de tipo MCAR

#### 2. Tipo MAR

Son ocasionados en fechas controladas como por ejemplo cuando se hace mantenimiento a las herramienas. La localizacion de los valores faltantes depende de otros valores observados

```python
(
    diabetes_df
    .missing.sort_variables_by_missingness()
    .sort_values(by='blood_pressure') # ordena los valores segun una columna
    .pipe(missingno.matrix)
)
```

Para este caso, los valores faltantes de la variable blood_pressure pueden estar relacionados a algunos valores de insulin o de skin_thickness por lo que son de tipo MAR

#### 3. Tipo MNAR

Este tipo de datos faltantes proviene de los limites que presentan las herramientas al tratar de hacer seguimiento fuera de su rango de medicion. La localizacion de los valores faltantes depende de los valores faltantes en si mismo

```python
(
    diabetes_df
    .missing.sort_variables_by_missingness()
    .sort_values(by='insulin') # ordena los valores segun una columna
    .pipe(missingno.matrix)
)
```

Vemos que hay un rango de insulina en la que estan la mayoria de los valores faltantes de las demas variables por lo que concluimos que es de tipo MNAR 

No podemos tener seguridad sobre cual mecanismo incide sobre los datos para generar los valores faltantes, pero a traves de analisis y conocimiento del tema se pueden hacer suposiciones razonables

Para este caso, vamos a crear una matriz de sombra con los datos del dataframe riskfactor

#### Construccion de la matriz de sombra manualmente

```python
(
    riskfactors_df
    .isna()                     # crea una matriz de booleanos con los datos del df
    .replace({                  # reemplaza los valores booleanos por valores informativos
        False:'Not missing',
        True: 'Missing'
    })
    .add_suffix('_NA')          # Agrega un sufijo sobre cada variable
    .pipe(
        lambda shadow_matrix: pd.concat( # concatena los valores de la matriz de sombra a la derecha del df
            [riskfactors_df, shadow_matrix],
            axis='columns'
        )
    )
)
```

## Visualizacion de valores faltantes en dos variables 

Permiten visualizar y compara dos variables cuando ambas le faltan valores. Un grafico muy util para ello, son los Scatterplot. 

Vamos a crear una funcion que agregue valores aleatorios a las variables con valores faltantes 

```python
'''
Funcion que agrega valores aleatorios a las variables con valores faltantes
para visualizarlos en un eje
'''

def column_fill_with_dummies(
    column: pd.Series,
    proportion_below: float=0.10, # proporcion de los datos en la grafica
    jitter: float=0.075,          # parametro que evita el asolapamiento de los puntos de la grafica
    seed: int=42                  # semilla para la aleatoriedad
) -> pd.Series:                   # la funcion retorna una Serie

    # copiar las columnas del DataFrame
    column = column.copy(deep=True)

    # extraer los valores de las variables 
    missing_mask = column.isna()                # matriz de booleanos
    number_missing_values = missing_mask.sum()  # conteo de valores faltantes
    column_range = column.max() - column.min()  # rango de las variables

    # Datos de turno
    column_shift = column.min() - column.min()*proportion_below

    # Ruido alrededor de los puntos
    np.random.seed(seed)
    column_jitter = (np.random.rand(number_missing_values) - 2) * column_range * jitter

    # guarda los nuevos datos aleatorios
    column[missing_mask] = column_shift + column_jitter

    return column
```

#### Scatterplot para dos variables con valores faltantes

```python
# analisis para las variables numericas Peso y Altura
(
    riskfactors_df
    .select_dtypes(
        exclude='category'  # excluye las variables categoricas
    )
    .pipe(                  # selecciona solo columnas que tengan valores faltantes
        lambda df: df[df.columns[df.isna().any()]]
    )                       # matriz de sombra con valor True para valores faltantes y False para valores completos
    .missing.bind_shadow_matrix(true_string = True, false_string = False)
    .apply(                 # funcion que rellena los valores faltantes con column_fill_with_dummies
        lambda column: column if '_NA' in column.name else column_fill_with_dummies(column, proportion_below=0.05, jitter=0.075)
    )                       
    .assign(                # nueva columna con las variables a graficar (en formato booleano)
        nullity = lambda df: df.weight_lbs_NA | df.height_inch_NA
    )
    .pipe(                  # funcion para graficar el scatterplot
        lambda df: (
            sns.scatterplot(
                data = df,
                x='weight_lbs',
                y='height_inch',
                hue='nullity'
            )
        )
    )
)
```

Los puntos azules de la grafica representan los valores completos, existen observaciones para ambas variables. Los puntos naranja indican que existe observacion para una variable y no para la otra. Para el caso de la altura vemos que existe una mayor cantidad de valores que tienen altura pero no el peso y que se agrupan en un rango que de 60 a 70 pulgadas. Para la variable Peso, vemos que solo tiene un valor que no contiene la Altura.

## Eliminacion de Valores Faltantes

Existen dos metodos para el tratamientos de valores faltantes: Eliminacion e Imputacion.

- En la eliminacion de valores faltantes se asume que los valores estan perdidos completamente al azar (MCAR). En cualquier otro caso, realizar una eliminacion podra ocasionar sesgos en los analisis y modelos subsecuentes.

- La imputacion consiste en a√±adir valores en los datos faltantes para mantener el registro en el datasets.

## Imputacion Basica de los Datos

Consiste en reemplazar los valores faltantes por algun valor, bien sea estadistico, algun valor por contexto o valores obtenidos mediante un modelo de Machine Learning 

### Imputacion con base en el contexto

implica conocer la estructura de los datos por lo que podemos completar las entradas vacias 

### Imputacion de un unico valor

Se da cuando el contexto no ayuda a completar un valor faltante y por tanto requieres imputar un unico valor con algun estadistico. Para este caso, utilizamos las variables Peso, Altura y bmi para aplicar el valor promedio a cada valor faltante. Ademas creamos la matriz de sombras de las tres variables para luego crear una grafica representativa 

```python
# Matriz de sombra para las variables Peso, Altura y bmi
(
    riskfactors_df                                          # dataframe
    .select_columns('weight_lbs', 'height_inch', 'bmi')     # variables seleccionadas
    .missing.bind_shadow_matrix(true_string=True, false_string=False) # creacion de la matriz de sombra
    .apply(
        axis = 'rows',
        func = lambda column: column.fillna(column.mean() if '_NA' not in column.name else column)
    )
)
```

#### Imputacion de Valores sobre Multiples Variables

```python
# Imputacion simultanea para las variables Peso, Altura y bmi
# Creamos una nueva columna con True para registros donde hayan valores faltantes
# en peso o altura
(
    riskfactors_df                                          # dataframe
    .select_columns('weight_lbs', 'height_inch', 'bmi')     # variables seleccionadas
    .missing.bind_shadow_matrix(true_string=True, false_string=False) # creacion de la matriz de sombra
    .apply(
        axis = 'rows',
        func = lambda column: column.fillna(column.mean() if '_NA' not in column.name else column)
    )  
    .assign(                                                # nueva columna
        imputed=lambda df: df.weight_lbs_NA | df.height_inch_NA
    )   
)
```

```python
# Imputacion simultanea para las variables Peso, Altura y bmi
(
    riskfactors_df                                          # dataframe
    .select_columns('weight_lbs', 'height_inch', 'bmi')     # variables seleccionadas
    .missing.bind_shadow_matrix(true_string=True, false_string=False) # creacion de la matriz de sombra
    .apply(
        axis = 'rows',
        func = lambda column: column.fillna(column.mean() if '_NA' not in column.name else column)
    )  
    .assign(                                                # nueva columna
        imputed=lambda df: df.weight_lbs_NA | df.height_inch_NA
    )  
    .pipe(                                                  # seccion de la grafica
        lambda df: (
            sns.scatterplot(
                data=df,
                x='weight_lbs',
                y='height_inch',
                hue='imputed'
            )
        )
    )
)
```

Los valores naranjas para este caso, ya no se encuentran fuera de la grafica sino que adquieren un valor promedio en ambas variables. Tales valores son valores imputados y son considerados dentro del conjunto total de datos.

- Si los valores imputados rompen la estructura de los datos implica que la imputacion sesga los datos y podria alterar los resultados. En este caso no altera la estructura de los datos

```python
# Histogramas para varias variables 
# Colocar las columnas variable y value como index
# crear nueva columna con true para los registros donde los valores de variables esten en variable_NA
# es decir, que exista el valor
(
    riskfactors_df                                          # dataframe
    .select_columns('weight_lbs', 'height_inch', 'bmi')     # variables seleccionadas
    .missing.bind_shadow_matrix(true_string=True, false_string=False) # creacion de la matriz de sombra
    .apply(
        axis = 'rows',
        func = lambda column: column.fillna(column.mean() if '_NA' not in column.name else column)
    )
    .pivot_longer(                                          # pivotear la matriz de sombra hacia los valores NA
        index='*_NA'                                        
    )
    .pivot_longer(
        index=['variable', 'value'],
        names_to='variable_NA',
        values_to='value_NA'
    )
    .assign(                                                # nueva columna 
        valid=lambda df: df.apply(axis='columns', func = lambda column: column.variable in column.variable_NA)
    )
    .query('valid')
)
```
