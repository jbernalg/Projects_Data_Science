# __Anlisis de personalidad de clientes (Parte I)__

Es un anlisis detallado que se realiza a los clientes para encontrar el cliente ideal de una empresa. Permite a las empresas comprender mejor a sus clientes para facilitar la adaptacin de sus productos segn las necesidades, comportamientos e inquietudes especficas de los diferentes tipos de clientes.

La forma en que lo vamos a realizar es segmentando los clientes de acuerdo a caractersticas comunes y encontrar el grupo que tiene mayor probabilidad de comprar un producto para luego comerciarlizarlo nicamente a ese segmento en particular, reduciendo costos de distribucin, marketing y ahorrando tiempo.

## __Contexto de los Datos__

Vamos a trabajar con un set de datos de clientes de una empresa. La descripcin de la informacin recolectada es la siguiente:

Informacin bsica: 

- _ID_: Identificador nico de cliente.
- _Year_Birth_: Ao de nacimiento del cliente.
- _Education_: Nivel educativo.
- _Marital_Status_: Estado civil.
- _Income_: Ingresos familiares anuales.
- _Kidhome_: Nmero de nios en el hogar.
- _Teenhome_: Nmero de adolescentes en el hogar.
- _Dt_Customer_: Fecha de alta del cliente en la empresa.
- _Recency_: Nmero de das desde la ltima compra del cliente.

Informacin de productos:

- _MntWines_: Cantidad gastada en vino en los ltimos 2 aos.
- _MntFruits_: Cantidad gastada en frutas en los ltimos 2 aos.
- _MntMeatProducts_: Cantidad gastada en carnes en los ltimos 2 aos.
- _MntSweetProducts_: Cantidad gastada en dulces en los ltimos 2 aos.
- _MntGoldProds_:Cantidad gastada en oro en los ltimos 2 aos.

Informacin de promociones tomada por los clientes:

- _NumDealsPurchases_: Nmero de compras realizadas con descuento.
- _AcceptedCmp1_: 1 si el cliente acept la oferta en la primera campaa, 0 en caso contrario.
- _AcceptedCmp2_: 1 si el cliente acept la oferta en la segunda campaa, 0 en caso contrario.
- _AcceptedCmp3_: 1 si el cliente acept la oferta en la tercera campaa, 0 en caso contrario.
- _AcceptedCmp4_: 1 si el cliente acept la oferta en la cuarta campaa, 0 en caso contrario.
- _AcceptedCmp5_: 1 si el cliente acept la oferta en la quinta campaa, 0 en caso contrario.
- _Response_: 1 si el cliente acept la oferta en la ltima campaa, 0 en caso contrario.

Informacin sobre el sitio de compras:

- _NumWebPurchases_: Nmero de compras realizadas a travs del sitio web de la empresa.
- _NumCatalogPurchases_: Nmero de compras realizadas a travs del catalogo.
- _NumStorePurchases_: Nmero de compras realizadas directamente en tiendas.
- _NumWebVisitsMonth_: Nmero de visitas al sitio web en el ltimo mes.


Informacin extra:
- _Complain_: Nmero de quejas dejada por el cliente.
- _Z_CostContact_: Costo por obtener al cliente.
- _Z_Revenue_: Ganancia por clic.
- _Response_: Respuesta.

### 6. Calcular la edad de los clientes

```python
# Convertimos la columna Dt_customer a tipo fecha
df['Dt_Customer'] = pd.to_datetime(df.Dt_Customer, format='%d-%m-%Y')
```

Obtenemos la fecha ms reciente de registro. Se asume que esa es la ltima fecha del sistema o el da de corte para calcular la antiguedad de cada cliente.

```python
last_date = df['Dt_Customer'].max()
```

Calculamos los das como cliente

```python
df['Days_is_client'] = (last_date - df['Dt_Customer']).dt.days
```

Calculamos la edad del cliente

```python
df['Client_age'] = (last_date.year - df['Year_Birth'])
df
```

Eliminamos las columnas que ya no necesitamos

```python
df.drop(columns = ['Year_Birth', 'Dt_Customer'], inplace=True)
df.head()
```

### 7. Redefinir las variables categoricas

Analicemos las categoras de cada variable categrica.

```python
for col in (df.select_dtypes('object').columns.tolist()):
    valores_unicos = df[col].value_counts()
    print(valores_unicos)
    print('\n')
```

Vamos a reestructurar los niveles educativos en tres: No graduados, graguados y con posgrado. Lo mismo harmos para el estado civil: En pareja y soltero.

```python
df['Education'].replace({
    'PhD':'Postgraduate',
    'Master':'Postgraduate',
    'Graduation':'Graduate',
    '2n Cycle':'Graduate',
    'Basic':'Undergraduate'
}, inplace=True)

df['Marital_Status'].replace({
    'Married':'Partner',
    'Together':'Partner',
    'Single':'Single',
    'Divorced':'Single',
    'Widow':'Single',
    'Alone':'Single',
    'Absurd':'Single',
    'YOLO':'Single'
}, inplace=True)
```

Verificamos que la conversin se haya hecho correctamente.

```python
for col in (df.select_dtypes('object').columns.tolist()):
    valores_unicos = df[col].unique()
    print(valores_unicos)
```

### 12. Categorizar variables en tres grupos basado en sus tipos de datos

```python
# variables con dos valores unicos
binary_columns = [col for col in df.columns if df[col].nunique() == 2]

# variables con mas de 2 y menos de 10 categorias
categorical_columns = [col for col in df.columns if 2 < df[col].nunique() < 10]

# variables numericas continuas
numerical_columns = [col for col in df.select_dtypes(include=['number']).columns if col not in binary_columns + categorical_columns]
```

### 13. Detectando outliers

Vamos a usar como criterio el Z-Score, que es una medida estadstica que indica cuntas desviaciones estndar se aleja un valor respecto a la media de su variable. 

```python
from scipy.stats import boxcox, zscore

# calcular Z-Score para cada valor numerico
z_score = pd.DataFrame(zscore(df[numerical_columns]), columns=numerical_columns)
z_score
```

```python
# identificar los registros donde cualquiera de las z-score supere el umbral
outliers = z_score[(np.abs(z_score) > 3).any(axis=1)]
outliers
```

```python
# eliminar los registros con outliers del dataframe
df = df.drop(outliers.index)
df.shape
```

> Pasamos de 2216 a 2198 registros libres de outliers. Fueron eliminados 18 registros con valores faltantes.

## __EDA__

Vamos a generar un conjunto de grficos de histograma con curvas de densidad para cada columna numrica del dataframe.

```python
import seaborn as sns

# inicializar variable para posicionar cada histograma
x = 1

# crear figura
plt.figure(figsize=(20,15))
# grilla de graficos
for col in numerical_columns:
    plt.subplot(len(numerical_columns)//2, len(numerical_columns)//2, x)
    sns.histplot(data=df, x=col, kde=True, color='blue')
    plt.title(f'Histograma de {col}', pad=10, fontweight='bold', fontsize=10)
    plt.tight_layout()
    x += 1
plt.show()
```

> La distribucin de ingresos (Income) sigue una normal, luego de haber sido eliminados los valores atpicos. Esto sugiere que la mayora de los clientes ganan alrededor del ingreso promedio y menos clientes ganan significativamente mas o menos.

> Las distribuciones de das como cliente (Days_is_client) y dias desde la ltima compra (Recency) son muy uniformes. Esto indica que los clientes has estado con la empresa durante distintos perodos de tiempo y han interactuado recientemente con ella en una amplia gama de perodos de tiempo

> La ditribucin de gastos (Expenses) muestran una distribucin exponencial lo que significa que la mayora de los clientes tienen un gasto menor y este disminuye rpidamente a medida que aumenta el monto.

> La distribucin del nmero total de compras (TotalNumPurchases) sigue una distribucin binomial, lo que evidencia la existencia de comportamientos de compra comunes entre los clientes como realizar una cantidad especfica de compras.

Ahora, generamos un set de grficos de barras para visualizar la distribucin de columnas categricas y binarias del dataframe. 

```python
# paleta de colores
paleta_pers = ['#327D7C', '#E2504A', '#F0C808']

# inicializar variable para posicionar cada histograma
x = 1

# crear figura
plt.figure(figsize=(20,15))
# grilla de graficos de variables categoricas y binarias
for col in categorical_columns + binary_columns:
    plt.subplot(3,3,x)
    ax = sns.countplot(data=df, x=col, palette=paleta_pers)
    plt.title(f'Distribucion de {col}', pad=10, fontweight='bold', fontsize=12)
    ax.bar_label(ax.containers[0])
    plt.tight_layout()
    x += 1

plt.show()
```

> La mayora de los clientes son graduados, con hijos, tienen pareja, no se han quejado en los ltimos dos aos y nunca han aceptado oferta de campaa. Esto sugiere un perfil de cliente especfico al que atiende la empresa.  

Aplicamos un anlisis de correlacin sobre las variables numricas 

```python
paleta_pers = ["#E2504A", "#EDEBEC", "#327D7C"]

# calcular matriz de correlacion
corr_matrix = df.select_dtypes(include='number').corr()

# crear mapa de 
# fmt formato de numero
plt.figure(figsize=(10,10))
sns.heatmap(data=corr_matrix, annot=True, color=paleta_pers, cmap='coolwarm', fmt='.1g')
plt.show()
```

> Los ingresos, los gastos y el numero total de compras son las caractersticas ms correlacionadas lo que sugiere que un mayor ingreso esta estrechamente vinculado con un mayor gasto y un mayor nmero de compras.
