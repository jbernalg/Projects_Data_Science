# __Construccin de modelos y Resultados__

## Importar libreras

```python
# librerias para el dataframe y la visualizacion
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# librerias para la clusterizacion
import sklearn
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
```

## Cargar los datos

```python
# datos con valores originales
rfm = pd.read_csv('./data/rfm.csv')
rfm.head()
```

```python
# datos con valores escalados
rfm_df_scaled = pd.read_csv('./data/rfm_df_scaled.csv')
rfm_df_scaled.head()
```

## Construir Modelo K-Means

Este modelo requiere inicializarlo con un nmero de clustering definido. Como desconocemos el valor, vamos a probar con diferentes tcnicas para encontrar el K ptimo.

### _Encontrar el nmero ptimo de clusteres_

Vamos a implementar dos tcnicas muy comunes para encontrar el K adecuado.

__Mtodo del codo__

```python
# lista para guardar la inercia de cada K
inercia = []

# definir el rango de K a probar
K = range(2, 9)

# iteramos sobre el modelo para cada valor de K
for n_cluster in K:
    kmeans = KMeans(n_clusters=n_cluster, max_iter=50, n_init=10, random_state=42)
    kmeans.fit(rfm_df_scaled)
    inercia.append(kmeans.inertia_)

# visualizamos los resultados
plt.figure(figsize=(10, 8))
plt.plot(inercia)
plt.xlabel('Clusteres')
plt.ylabel('Inercia')
plt.show()
```

> El grfico muestra la mayor prdida de inercia en el cluster 1. Sin embargo, 1 slo cluster no nos aporta informacin relevante. 

> El cluster 3 presenta un punto de inflexin en cuanto a la prdida de inercia. Es un buen valor a considerar.

__ndice de Silueta__

```python
# lista para guardar los indice de silueta para cada cluster
index_silueta = []

# rango de K a probar
K = range(2, 9)

# itermaos el modelo para cada valor de K
for n_cluster in K:
    kmeans = KMeans(n_clusters=n_cluster, n_init=10, max_iter=50, random_state=42)
    y_pred = kmeans.fit_predict(rfm_df_scaled)
    index_silueta.append(silhouette_score(rfm_df_scaled, y_pred))

# visualizamos resultados
plt.figure(figsize=(10, 8))
plt.plot(index_silueta)
plt.xlabel('Clusteres')
plt.ylabel('Coeficiente de silueta')
plt.show()
```

> La grafica muestra en el cluster 3 el coeficiente de silueta no decae tanto.

> En ambos graficos la sugerencia es K=3, por lo que lo seleccionamos como el K ptimo.

### _Mejor Modelo K-Means_

Ambos mtodos sugieren que con un nmero de cluster igual a 3 podemos obtener un modelo de clusterizacin ptimo. 

```python
# definimos el modelo
kmeans = KMeans(n_clusters=3, n_init=10, max_iter=50, random_state=42)

# entrenamos y obtenemos las predicciones
y_predict = kmeans.fit(rfm_df_scaled)

# asignamos la etiqueta de clusterizacion a cada cliente del df
rfm['Cluster_K'] = kmeans.labels_
rfm.head() 
```

Vamos a obtener el ndice de silueta del modelo seleccionado para realizar comparaciones al final y decidir cul es el mejor a partir de esta mtrica.

```python
print(silhouette_score(rfm_df_scaled, y_pred))
```

> El indice de silueta general del modelo es bueno.

## Construir modelo Hierarchical Clustering

Vamos a utilizar el modelo de aglomeracin jerrquica para clusterizar los clientes de acuerdo a RFM. Vamos a probar con diferentes mtodos hasta encontrar aquel con el mejor resultado y de esta forma etiquetar a cada cliente.

### _Single Linkage_

la distancia entre dos grupos se define como la distancia ms corta entre dos puntos en cada grupo.

```python
fig = plt.figure(figsize=(10,10))
dendogram_plot = dendrogram(linkage(rfm_df_scaled, method='single', metric='euclidean'))
plt.xlabel('Clusteres')
plt.ylabel('Distancia euclideana')
plt.show()
```

> No muestra claridad en cuanto a los clusteres que se pueden formar.

> Logra percibirse dos grupos: azul y verde, este ltimo siendo ms predominante.

### _Complete linkage_

La distancia entre dos grupos se define como la distancia ms larga entre dos puntos en cada grupo.

```python
fig = plt.figure(figsize=(10,10))
dendogram_plot = dendrogram(linkage(rfm_df_scaled, method='complete', metric='euclidean'))
plt.xlabel('Clusteres')
plt.ylabel('Distancia euclideana')
plt.show()
```

> Se logra diferenciar 3 clusteres

> 2 de los clusteres son muy dominantes respecto al tercero.

> Arroja mejores resultados que el dendograma anterior.

### _Average linkage_

La distancia entre dos grupos se define como la distancia promedio entre cada punto de un grupo y cada punto del otro grupo.

```python
fig = plt.figure(figsize=(10,10))
dendogram_plot = dendrogram(linkage(rfm_df_scaled, method='average', metric='euclidean'))
plt.xlabel('Clusteres')
plt.ylabel('Distancia euclideana')
plt.show()
```

> Se evidencia la diferencia entre dos clusteres.

> No es tan claro pero se evidencia hasta 3 clusteres.

> Hay un cluster dominante frente al resto.

### _Warm_

Realiza un monton de operaciones en el que cada dato se lleva a otro cluster, evala su varianza y se queda con la varianza que de menor.

```python
fig = plt.figure(figsize=(10,10))
dendogram_plot = dendrogram(linkage(rfm_df_scaled, method='ward', metric='euclidean'))
plt.xlabel('Clusteres')
plt.ylabel('Distancia euclideana')
plt.show()
```

> Se evidencia una mejor separacin entre 3 clusteres.

> Los 3 clusteres mantienen una buena distancia entre s, lo que indica una excelente clusterizacin.

### _Mejor Modelo Hierarchical Clustering_

De los mtodos probados, nos quedamos con Warm debido a su excelente diferenciacin de clusteres. Vamos a calcular el ndice de silueta general para 2 y 3 clusteres y as encontrar la mejor opcin.

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# para 2 clusteres
hc_2 = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
y_hc_2 = hc_2.fit_predict(rfm_df_scaled)
print(silhouette_score(rfm_df_scaled, y_hc_2))
```

```python
# para 3 clusteres
hc_3 = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
y_hc_3 = hc_3.fit_predict(rfm_df_scaled)
print(silhouette_score(rfm_df_scaled, y_hc_3))
```

> Evidentemente, con 2 cluster obtenemos un ndice de silueta mucho mayor debido a que dos de clusteres estan bien distanciados. Sin embargo, diferenciar los clientes en 2 grupos no nos brinda la informacion que necesitamos.

> Para 3 clusteres obtenemos un ndice de silueta bastante bueno y 3 grupos de clientes si podria ser relevante en cuanto a la creacion de campaas.

> Nos quedamos con la clusterizacion para K=3 utilizando el mtodo Warm. Guardamos la etiqueta de cada cliente en el dataframe

```python
rfm['cluster_hc'] = y_hc_3
rfm.head()
```

## Construir modelo DBSCAN

Este modelo requiere determinar un rango de valores de epsilon (eps) y un rango de nmero mnimo de puntos vecinos dentro del radio eps. Procedemos a calcular cada rango.

### Rango ptimo de epsilon

```python
from sklearn.neighbors import NearestNeighbors

# calculo del vecino mas cercano
neighbors = NearestNeighbors(n_neighbors=2)
# ajuste del objeto a los datos
neighbors_fit = neighbors.fit(rfm_df_scaled)
# guardamos la distancia e indice de cada punto
distances, indices = neighbors_fit.kneighbors(rfm_df_scaled)
```

```python
# ordenamos las distancias
distances = np.sort(distances, axis=0)
# tomar la distancia del vecino mas cercano de cada dato
distances = distances[:,1]
```

Creamos el grfico de k-distancias donde se ordenan la distancias al k-simo vecino ms cercano. El codo generado indica un buen rango de valores para eps.

```python
plt.figure(figsize=(12, 8))
plt.plot(distances)
plt.ylabel("Distancia al {}-simo vecino")
plt.xlabel("Puntos ordenados")
plt.title("Grfico k-distancia para estimar eps")
plt.grid(True)
plt.show()
```

> El codo se encuentra entre 0.10 y 0.55 por lo que elegimos ese rango como rango de epsilon

```python
# rango de epsilon
eps_values = np.arange(0.10, 0.60, 0.05)
```

### Rango ptimo de min_samples

Para definir este rango, vamos a utilizar una regla emprica que indica que el valor mnimo es 3 y el valor mximo es el doble del mnimo. Sin embargo, es importante tener en cuenta que este rango se suele ajustar mejor por ensayo y error. Para este caso, lo definimos entre 3 y 7:

```python
# rango de min_samples
min_samples = np.arange(3, 8)
```

### Obtener predicciones del modelo DBSCAN con las diferentes combinaciones de parmetros

```python
# libreria para combinar parametros
from itertools import product

# combinatoria de parametros
dbscan_params = list(product(eps_values, min_samples))

# lista para almacenar el numero de cluster
n_cluster = []

# lista para almacenar el indice de silueta
sil_score = []

# iteramos el modelo con la diferentes combinaciones de parametros
for p in dbscan_params:
    y_dbscan = DBSCAN(eps=p[0], min_samples=p[1]).fit_predict(rfm_df_scaled)
    
    try:
        sil_score.append(silhouette_score(rfm_df_scaled, y_dbscan))
    except:
        sil_score.append(0)

    n_cluster.append(len(np.unique(y_dbscan)))
```

Creamos un datarame con las predicciones del modelo junto a sus correspondientes parmetros.

```python
df_param_tunning = pd.DataFrame.from_records(dbscan_params, columns=['Eps', 'Min_samples'])
df_param_tunning['Sil_score'] = sil_score
df_param_tunning['N_clusters'] = n_cluster

df_param_tunning.head()
```

Creamos dos tablas dinmicas a partir del dataframe anterior para visualizar mejor el rendimiento de las distintas combinaciones de parmetros en DBSCAN.

```python
pivot_1 = pd.pivot_table(df_param_tunning, values='Sil_score', columns='Eps', index='Min_samples')
pivot_2 = pd.pivot_table(df_param_tunning, values='N_clusters', columns='Eps', index='Min_samples')

pivot_1
```

### Heatmap del ndice de silueta y del nmero de clusteres

```python
# heatmap del indice de silueta
fig, ax = plt.subplots(figsize=(18,6))
sns.heatmap(pivot_1, annot=True, annot_kws={'size':10}, cmap='coolwarm', ax=ax)
plt.show()
```

```python
# heatmap del numero de cluster
fig, ax = plt.subplots(figsize=(18,6))
sns.heatmap(pivot_2, annot=True, annot_kws={'size':10}, cmap='coolwarm', ax=ax)
plt.show()
```

> Comparando ambos heatmaps, vemos que el ndice de silueta ms alto (0.55) corresponde a un total de 3 clusteres que no es el ms bajo. Por tanto, esta eleccin se considera buena y corresponde a los parmetros eps=0.55 y min_sample=7

> Vamos a entrenar DBSCAN con los parmetros seleccionados para contrastar el resultado que se obtenga con el ndice de silueta dado en el heatmap.

```python
dbscan_train = DBSCAN(eps=0.55, min_samples=7)
y_dbscan = dbscan_train.fit_predict(rfm_df_scaled)
print(silhouette_score(rfm_df_scaled, y_dbscan))
```

> Como vemos, el indice de silueta que arroja el modelo concuerda con el valor presentado en el heatmap.

> Guardamos las predicciones de DBSCAN en el dataframe con los datos originales 'rfm'

```python
rfm['Cluster_dbscan'] = y_dbscan
rfm.head()
```
