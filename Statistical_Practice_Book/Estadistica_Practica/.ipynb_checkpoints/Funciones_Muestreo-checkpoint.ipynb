## 1) Muestreo Aleatorio Simple
Metodo de seleccion de ciertas unidades sacadas de una poblacion de manera que cada una de las muestras tiene la misma probabilidad de ser elegida. Ejemplo, la loteria

```python
# muestra de 8 registros
aleat_8 = econdata.sample(n=8)
aleat_8
```

```python
# creamos otra muestra de 8 elementos para corroborar que funciona correctamente
aleat_8_2 = econdata.sample(n=8)
aleat_8_2
```

```python
# muestra del 25% del total de la poblacion
prop_25 = econdata.sample(frac=.25)
prop_25.head()
```

## 2) Muestreo Sistematico
Metodo de seleccion de ciertas unidades al azar y, a continuacion, se eligen el resto siguiendo intervalos regulares. Se imponen condiciones internas o externas para la seleccion. Ejemplo, dar un premio cada cien personas que hagan una inscripcion hasta llegar a un total de mil inscritos

```python
# creamos una funcion que nos permita elegir los intervalos de eleccion de la muestra
def systematic_sampling(econdata, step):
    indexes = np.arange(0, len(econdata), step=step)
    systematic_sample = econdata.iloc[indexes]
    return systematic_sample
```

```python
# creamos la muestra tomando los registros de 3 en 3
systematic_sample = systematic_sampling(econdata, 3)
systematic_sample.head(10)
```

```python
# otra alternativa a la funcion
def system_sampling_2(data, start, end, step):
     return data.iloc[start:end:step,:]
```

```python
# muestra de registros de 10 en 10 en los primeros 200 
system_sampling = system_sampling_2(econdata, 0, 200, 10)
system_sampling
```

## 3) Muestreo Estratificado
Metodo de seleccion de ciertas unidades por segmentos exclusivos y homogeneos y, a continuacion, se elige una muestra aleatoria simple de cada segmento. Ejemplo, division de la muestra por edades, region o pais de las personas

```python
# realizamos un conteo de las categorias de la variable estratificado
# expresamos proporcion de cada categoria en %
# ordenamos las categorias de mayor a menor segun su porcentaje
(econdata['estratificado'].value_counts(normalize=True).sort_values(ascending=False))
```

Imaginemos que los datos de la poblacion total indican las proporciones siguientes:

1) Hoteles en Cuauhtemoc: 0.5

2) Museos en Cuauhtemoc: 0.2

3) Hoteles en Venustiano Carranza: 0.1

4) Mercados en Cuauhtemoc: 0.1

5) Mercados en Venustiano Carranza: 0.1

### 3. Toma tantos datos como necesites de cada categoría para llegar a la proporción deseada.

### 4. Junta esas muestras de cada categoría para tu muestreo final.

Debemos adaptar las proporciones de la muestra con las dadas por la poblacion para que los resultados sean compatibles con la realidad. Para ello, vamos a crear una funcion que nos arroje un nuevo DataFrame con una muestra estratificada de los datos originales

```python
# funcion 1
def data_estratificada(econdata, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state=None):

  df_estrat = pd.DataFrame(columns = econdata.columns) 

  pos = -1
  for i in range(len(valores_estrat)):
    pos += 1
    if pos == len(valores_estrat) - 1: 
      ratio_len = len(econdata) - len(df_estrat)
    else:
      ratio_len = int(len(econdata) * prop_estrat[i])

    df_filtrado = econdata[econdata[nombres_columnas_estrat] ==valores_estrat[i]]
    df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state)

    df_estrat = pd.concat([df_estrat, df_temp]) 
  return df_estrat
```

```python
# nombre de las categorias
valores_estrat = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel', 'Cuauhtémoc,Mercado','Venustiano Carranza,Mercado']
# proporcion deseada de las diferentes categorias
prop_estrat = [0.5, 0.2, 0.1, 0.1, 0.1]
# dataframe con la nueva muestra
df_estrat = data_estratificada(econdata, 'estratificado', valores_estrat, prop_estrat, random_state=42)
df_estrat.head()
```

```python
# mostrar las proporciones de la nueva muestra
(df_estrat['estratificado'].value_counts()/len(df_estrat)).sort_values(ascending=False)
```

```python
# funcion 2
def Estratificado(datos, columna, size, valores_estrat, random_state = None):
    
    # dataframe vacio con los nombres de las columnas del df dado  
    df_estrat = pd.DataFrame(columns=datos.columns)
  
    # obtener las categorias de la columna estratificada
    datos_en_col = datos[columna].unique()

    con = 0
    
    # recorrer cada categoria
    for i in datos_en_col :
        # en un nuevo df, guardar los registros de cada categoria
        df_filtrado = datos[datos[columna] == datos_en_col[con]]
        # obtener una muestra de tamano n correspondiente a la proporcion dada en valores_estrat
        # para la categoria seleccionada
        df_temp = df_filtrado.sample(replace=True , n= int(valores_estrat[con] *size) ,random_state= random_state)
        # guardamos en df_strat las muestras de cada categoria
        df_estrat = pd.concat([df_estrat,df_temp])
        # sumamos 1 a la posicion del contados
        con += 1
    
    return df_estrat
```

```python
# proporcion deseada de las diferentes categorias
val_estrat = [0.5,0.2,0.1,0.1,0.1]
# dataframe con la nueva muestra 
df_p = Estratificado(econdata,'estratificado',150,val_estrat, random_state=4)
# proporcion de cada categoria en la nueva muestra
print((df_p['estratificado'].value_counts(normalize=True)).sort_values(ascending=False))

df_p.head()
```
