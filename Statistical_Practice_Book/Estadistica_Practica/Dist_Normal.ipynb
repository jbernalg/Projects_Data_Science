# Todo sobre la Distribución Normal

Es una distribución icónica en la estadística que se presenta en muchos fenómenos naturales y procesos aleatorios como encuestas. El término Normal indica que es una de las más frecuentes y útiles en la estadística, no que sea la única o la más correcta.

Veamos un ejemplo para ir comprendiendo de qué se trata

### *Tenemos un grupo de personas de una ciudad y queremos estudiar la distribución de la altura y la media de toda la población*

Vamos midiendo cada persona y los resultados los vamos organizando en un histograma

La grafica anterior se ajusta a una curva de distribución normal

Esta curva tiene algunas características importantes:

> - La altura media es también la más frecuente. La media es igual a la moda.
> - La distribución es simétrica respecto a la media. Hay tanta gente que mide mas de la media, como gente que mide menos de la media.
> - Muchas encuestas o parámetros a estudiar se ajustan a esta curva
> - La media poblacional se representa por $\mu$
> - La desviación típica, que se representa por $\sigma$, es una medida de qué tan cerca o lejos están los datos de la media $\mu$. Para una población heterogénea, $\sigma$ es muy alta; para una población homogénea, $\sigma$ es muy baja

## Función de Distribución Normal

Carl Friedrich Gauss, se propuso a encontrar una función matemática que se ajuste a la curva de la distribución normal y dedujo la siguiente fórmula:

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}}e^{-\frac{1}{2}\left ( \frac{x - \mu}{\sigma} \right )^{2}}$$

Las características de esta función son las siguientes:
- Considera la media ($\mu$) y la desviación típica ($\sigma$).
- Es simétrica respecto a la media, al igual que la distribución.
- Tiene una asíntota horizontal que es el eje x.
- El área entre la función y el eje horizontal es igual a 1. Recordemos que para calcular áreas bajo una curva, en matemáticas, utlizamos integrales:

$$Area = \int_{-\infty}^{\infty}\frac{1}{\sigma \sqrt{2\pi}}e^{-\frac{1}{2}\left ( \frac{x - \mu}{\sigma} \right )^{2}} dx = 1$$

Al valer toda el área 1, podemos extrapolar a que es el 100% de la población ya que toda la población esta englobada bajo esa función.

Veamos un ejemplo

### *Supongamos que en un determinado país la estatura de la población adultas sigue una distribución normal de media 170 cm y desviación típica igual a 12 cm*

La distribución normal, para el ejemplo, se representa de la siguiente forma:

$$N(\mu, \sigma)\rightarrow N(170, 12)$$

La función, para los valores de $\mu$ y $\sigma$ dados se representa como sigue:

$$ f(x)= \frac{1}{12 \sqrt{2\pi}}e^{-\frac{1}{2}\left ( \frac{x - 170}{12} \right )^{2}}$$

Si evaluamos la función para un rango contínuo entre 130 y 210 obtenemos la siguiente distribución:



Claramente vemos que en el centro esta el valor de 170, como indicaba la media dada y vemos como a medida que nos acercamos a estaturas extremas como 130 cm y 210 cm, el porcentaje de la población es cada vez mas pequeño.

### *Que porcentaje de esa población mide más de 170 cm?*

Este porcentaje de la población viene representado por lo que estan a la derecha de 170 que es la media



Entonces, para caclularlo debemos integrar la funcion desde 170 hasta infinito como sigue:

$$Poblacion(X > 170 ) = \int_{170}^{\infty}\frac{1}{12 \sqrt{2\pi}}e^{-\frac{1}{2}\left ( \frac{x - 170}{12} \right )^{2}} dx$$

Vamos a calcular la integral con la libreria Sympy de Python

```python
# libreria para calculo de integrales
import sympy as sp

# Definimos la variable x
x = sp.Symbol('x')

# definimos la funcion
f = (1/(12*sp.sqrt(2*sp.pi)))*sp.exp(-0.5*((x-170)/12)**2)

# calculo de la integral
result_integral = sp.integrate(f,(x, 170, sp.oo))

# mostrar resultado
result_integral
```

## Distribucion Normal Estandar

Es una distribucion normal con media de 0 y desviacion estandar 1, valores que hacen que la formula se simplifique considerablemente quedando de la siguiente forma:

$$f(x) = \frac{1}{\sqrt{2\pi}}e^{-\left ( \frac{x}{2} \right )^{2}}$$

Ayuda a que el calculo de area se simplifique al utilizar la tabla Z, la cual contiene tabulado los valores del area bajo la curva para un rango definido entre 0 y 3.90. Su forma es la siguiente:



De esta manera, ya no tenemos que acudir a integrales tediosas para calcular el area hasta un valor dado, sino que vamos a la tabla y ubicamos los valores de area asociados a dicho valor.

En lo que sigue vamos a crear la tabla Z con Python y calcular el area para ciertos valores

### *Tabla Z*

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

tabla_estandarizada = pd.DataFrame(
    [], 
    index=["{0:0.2f}".format(i / 100) for i in range(0, 400, 10)],
    columns = ["{0:0.2f}".format(i / 100) for i in range(0, 10)])

for index in tabla_estandarizada.index:
    for column in tabla_estandarizada.columns:
        Z = np.round(float(index) + float(column), 2)
        tabla_estandarizada.loc[index, column] = "{0:0.4f}".format(norm.cdf(Z))

tabla_estandarizada.rename_axis('Z', axis = 'columns', inplace = True)
tabla_estandarizada
```

## *Ejemplo 1: Dada una distribucion $N(0,1)$* 

### *a.- Qué porcentaje queda por debajo del valor 1?*
Para calcularlo, debemos integrar la formula de la distribucion normal estandar desde - infinito hasta 1. Sin embargo, como ya tenemos la tabla Z, podemos apoyarnos de ella para realizar el calculo. El area a calcular es la siguiente:



Los valores de Z se encuentran en la primera fila y columna. Alli ubicamos el valor hasta donde se desea calcular el area. Al encontrarlo, ubicamos el respectivo valor de area. Para este caso obtenemos:

$$P(X < 1) = 0.8413$$

El resultado indica que el 84.13% de la poblacion esta por debajo de 1 en esta distribucion.

### *b.- Que porcentaje queda por encima del valor 0.5?*
En este caso, debemos calcular el area que se encuentra entre 0.5 e infinito. 



Procedemos como sigue:

1.- Calculamos el valor de area de - infinito hasta 0.5, buscando en la tabla Z el area correspondiente a 0.5

$$P(X \leq 0.5) = 	0.6915$$

2.- Ahora, restamos al area total, que es de 1, el area calculada anteriormente

$$P(X > 0.5) = 1 - P(X \leq 0.5) = 1 - 0.6915 = 0.3085$$

El resultado indica que el 30.85% de la poblacion esta por encima de 0.5 en esta distribucion.

### *c.- Qué porcentaje queda entre -2 y 1?*
El area que debemos calcular va entre -2 y 1 y se representa como sigue:



Procedemos como sigue:

1.- Calculamos el area desde -infinito hasta 1.

$$P(X \leq 1) = 0.8413$$

2.- Calculamos el area desde -infinito hasta -2. Aqui se presenta un problema y es que **la tabla Z no contiene el area para valores negativos**. Sin embargo, podemos apoyarnos de la simetria de la distribucion para hacer el calculo. Si calculamos el area desde -infinito hasta 2 y restamos dicho valor al area total encontraremos el area de -infinito hasta -2, por simetria.

$$P(X \leq 2) = 0.9772$$

$$P(X \leq -2) = 1 - P(\leq 2) = 1 - 0.9772 = 0.0228$$

3.- Restamos a $P(X \leq 1)$ el area de $P(X \leq -2)$

$$P(-2 \leq X \leq 1) = P(X \leq 1) - P(X \leq -2) = 0.8413 - 0.0228 = 0.8185$$

El resultado indica que el 81.85% de la poblacion se encuentra entre -2 y 1.

### **d.- Que valor deja por encima de si al 25.14% de la poblacion?**

En este caso, el area viene representada por la siguiente imagen:



Como no podemos calcular el area por encima de un valor, lo que hacemos es encontrar el valor de $z$ para el area por debajo. Entonces al area total, restamos el area dada:

$$A(X < z) = 1 - A(X > z) = 1 - 0.2514 = 0.7486$$

Ahora, buscamos el valor de $z$ asociado a la area calculada. En la tabla verificamos que corresponde a un valor de $z$ de 0.67. Entonces $z = 0.67$ deja por encima al 25.14% de la poblacion.

## Tipificacion

Es una operacion en el que se transforma la variable de una distribucion normal en su equivalente en una distribucion $N(0,1)$ que permita utilizar la tabla $z$ y asi simplificar los calculos de areas bajo la curva. La formula para tipificar la variable es la siguiente:

$$z = \frac{x - \mu}{\sigma}$$

Entendamos las ventaja de la tipificacion mediante el siguiente ejemplo:

## *Ejemplo 2: En un determinado pais la estatura de la poblacion adulta sigue una distribucion $N(170, 12)$.*

### **a.- Que porcentaje de esa poblacion mide menos de 185 cm?**

Para este ejercicio la distribucion es de $\mu = 170$ y $\sigma = 12$ y nos pide calcular el porcentaje de la poblacion menor a 185 cm, $P(X \leq 185)$. La representacion de esta distribucion es la siguiente:



Podemos calcular el area mediante integrales. Sin embargo, para evitar calculos tediosos podemos convertir la distribucion normal en una distribucion normal estandar $N(0,1)$ que nos permita utilizar la tabla $z$ y calcular el area de forma rapida. Apliquemos la tipificacion:

$$z = \frac{x - \mu}{\sigma} = \frac{185 - 170}{12} = 1.25$$

Este valor de $z$, equivale al valor de $X$ para una distribucion $N(170, 12)$. Entonces, calcular la probabilidad de estar por debajo de 185 cm equivale a calcular la probabilidad de estar por debajo de 1.25 en una distribucion normal estandar:

$$P(X \leq  185) = P(z \leq 1.25)$$

Al buscar en la tabla encontramos que el valor de $z$ es igual a 0.8944. Esto indica que el 89.44% de la poblacion estan por debajo de 185 cm.

### **b.- Que porcentaje de esa poblacion mide mas de 149 cm?**

La probabilidad que nos piden calcular es $P(X > 149)$ y la representacion del area es la siguiente:



Al tipificar la variable obtenemos:

$$z = \frac{x - \mu}{\sigma} = \frac{149 - 170}{12} = -1.75$$

La probabilidad de estar por encima de 149, equivale en la tabla a la probabilidad de estar por encima de -1.75:

$$P(X > 149) = P(Z > -1.75)$$

Graficamente, la equivalencia se ve de esta manera:



Como en la tabla no podemos calcular $z$ negativas, entonces realizamos la siguiente operacion algebraica:

$$P(Z > -1.75) = P(Z \leq 1.75)$$

Al buscar en la tabla obtenemos una probabilidad de 0.9599. Entonces, el porcentaje de poblacion que mide mas de 149 cm es del 95.99%

### **c.- Que estatura deja por debajo de si al 80% de la poblacion?**

En este caso nos piden encontrar la $z$ que deja por debajo al 80% de la poblacion. Graficamente es calcular el area de la siguiente distribucion:



Para resolverlo, primero buscamos en la tabla el valor de $z$ que corresponde al 0.8. El valor mas proximo a 0.8 en la tabla es de 0.7995 que corresponde al valor de $z$ de 0.84

Ahora lo que hacemos es, mediante la formula de tipificacion, despejar $X$ y encontrar su valor

$$z = \frac{x- \mu}{\sigma} \rightarrow x = z\cdot\sigma + \mu $$
$$x = z\cdot\sigma + \mu = 0.84\cdot 12 + 170 = 180.08$$

El resultado indica que el 80% de la poblacion mide menos que 180,08 cm.

## *Ejemplo 3: El peso en gramos de las cajas de cereales de cierta marca sigue una distribucion $N(500,5)$*

### **a.- Calcula la probabilidad de encontrar una caja que pese menos de 496g (con librerias de Python)**

Vamos a utilizar la libreria de Python llamada Stats de Scipy con la que podemos calcular la distribucion normal sin necesidad de convertirla a una distribucion normal estandar y aprovechando el poder de computo  para agilizar el calculo.

```python
from scipy.stats import norm
```

Nos piden encontrar la probabilidad de encontrar una caja que pese menos de 496g,
$$P(X \leq 496)$$
Veamos como resolverlo

```python
# obtener parametros de la distribucion

# media
media = 500
# desviacion estandar
std = 5
# variable
x = 496
```

Como nos piden la probabilidad inferior a un valor de la variable, utilizamos la probabilidad acumulada desde -infinito hasta dicho valor (metodo cdf).

```python
prob = norm.cdf(x, media, std)
prob
```

> Hay una probabilidad del 21.19% de que salga en esa fabrica una caja que pese menos de 496g

### **b.- Que porcentaje de cajas pesa entre 505 y 510 gramos (con Python)?**

Para resolver, calculamos la probabilidad acumulada para ambos valores y luego se obtiene la diferencia de probabilidades:

$$P(505 < X < 510) = P(X \leq 510) - P(X \leq 505)$$

Graficamente, debemos calcular la siguiente area:



Veamos como resolverlo:

```python
# obtenemos parametros

#variable 1
X1 = 505
# variable 2
X2 = 510
```

```python
# probabilidad acumulada de X1
prob_1 = norm.cdf(X1, media, std)

# probabilidad acumulada de X2
prob_2 = norm.cdf(X2, media, std)

# probabilidad
prob = prob_2 - prob_1
prob
```

> La probabilidad que en esa fabrica una caja pese entre 505 y 510g es de 13.59%

## *Ejemplo 4: En una ciudad se estima que la temperatura maxima en el mes de junio sigue una distribucion normal, con media 23 grados y desviacion tipica de 5 grados. Calcular el numero de dias del mes en los que se espera alcanzar maximas entre 21 y 27 grados*

Para calcular el numero de dias, se procede como sigue:
- Calculamos la probabilidad, con la distribucion normal dada, de ambos limites 
- Se obtiene la diferencia de probabilidad entre ambos limites la cual representa la probabilidad en dicho rango. 
- Multiplicas la probabilidad del rango por los dias del mes de junio y de esta manera obtienes la cantidad de dias que buscas

```python
# obtener parametros de la distribucion

# media
media = 23
# desviacion tipica
std = 5
```

```python
# calcular probabilidad acumulada para limite inferior
prob_21 = norm.cdf(21, media, std)
# calcular probabilidad acumulada para limite superior
prob_27 = norm.cdf(27, media, std)

# probabilidad de la D. normal en dicho rango
prob = prob_27 - prob_21
prob
```

Ahora, solo queda multiplicar la probabilidad encontrada por la cantidad de dias al mes para obtener los numeros de dias donde se espera alcanzar maximas temperaturas.

```python
n_dias = 30 * prob
n_dias = n_dias.round(0)
n_dias
```

> La cantidad de dias donde se espera alcanzar la maxima de temperatura son 13.

## *Ejemplo 5: La media de los pesos de 500 estudiantes es 70 kg y la desviacion tipica es 3 kg. Suponiendo que los pesos se distribuyen normalmente, hallar cuantos estudiantes pesan:*

### **a.- Entre 60 y 75 kg**
Vamos a calular la probabilidad para ambos limites, con la distribucion normal dada y luego obtenemos la diferencia entre ambas probabilidades

```python
# obtener parametros

# media
media = 70
# desviacion tipica
std = 3
# total alumnos
total = 500
```

```python
# probabilidad limite inferior
prob_60 = norm.cdf(6, media, std)
# probabilidad limite superior
prob_75 = norm.cdf(75, media, std)
# probabilidad asociada a dicho rango
prob = prob_75 - prob_60
prob
```

Ahora multiplicamos la cantidad de alumnos por la probabilidad del rango para obtener el numero de alumnos.

```python
total_a = total * prob
total_a = total_a.round(0)
total_a
```

   > El numero de estudiantes que pesan entre 60 y 75kg es de 476.

### **b.-  Mas de 90 kg**

Calculamos la probabilidad acumulada hasta 90 y luego restamos a la probabilidad total el valor de probabilidad calculado. 

$$P(X > 90) = 1 - P(X \leq 90)$$

Luego, se encuentra la cantidad de estudiantes que pesan mas de 90kg multiplicando la cantidad total de estudiantes por la probabilidad de mas de 90, $P(X \leq 90)$

```python
# probabilidad acumulada hasta 90
prob_90 = norm.cdf(90, media, std)

# probabilidad para mas de 90
prob_mayor_90 = 1 - prob_90

# numero de estudiantes que pesan mas de 90
num_a_mayor_90 = total*prob_mayor_90
num_a_mayor_90 = num_a_mayor_90.round(0)
num_a_mayor_90
```

> No existe estudiante que pese mas 90kg

### **c.- Menos de 64 kg**
Calculamos la probabilidad acumulada hasta 64 y luego multiplicamos dicha probabilidad por la cantidad total de estudiantes para asi hallar la cantidad de aquellos que pesan menos de 64 kg.

```python
# probabilidad acumulada hasta 64
prob_64 = norm.cdf(64, media, std)

# numero de estudiantes que pesan menos de 64
total_a_menos_64 = prob_64*total
total_a_menos_64 = total_a_menos_64.round(0)
total_a_menos_64
```

> Hay 11 estudiantes que pesan menos de 64 kg

### **d.- Exactamente 64 kg**

Calculamos la probabilidad de 64 y luego multiplicamos por la cantidad total de estudiantes

```python
# probabilidad de obtener 64 con la funcion de probabilidad (pdf)
prob_64_exacto = norm.pdf(64, media, std)

# numero de estudiantes que pesan 64 
total_64 = prob_64_exacto*total
total_64 = total_64.round(0)
total_64
```

> Solo 9 estudiantes pesan exactamente 64 kg.

## *Ejemplo 8: En un examen tipo test de 200 preguntas de eleccion multiple, cada pregunta tiene una respesta correcta y una incorrecta. Se aprueba si se contestan mas de 110 respuestas correctamente. Suponga que se contesta al azar, calcular la probabilidad de aprobar el examen*

El ejemplo, claramente trata sobre una distribucion binomial ya que solo existen dos opciones posibles: respuesta correcta e incorrecta. Cada eleccion es independiente de las demas y la probabilidad de acertar una pregunta es del 50%.

```python
# obtener parametros de la distribucion binomial

# probabilidad de exito
p = 0.5
# probabilidad de fracaso
q = 0.5
# numero de eventos
n = 200
```

La distribucion binomial se expresa como sigue:

$$B(n,p) \rightarrow B(200, 0.5)$$

Debido a la gran cantidad de eventos, podemos aproximar esta distribucion a una normal solo si cumple con los siguientes requisitos:

$$np > 5$$
$$nq > 5$$

Verifiquemos que se cumplan las condiciones:

```python
print(n*p)
print(n*q)
```

Como cumple con las condiciones podemos hacer la aproximacion:

$$B(n,p) \sim N(np,\sqrt{npq}) \rightarrow N(100, 7.07)$$

Ahora, podemos calcular la probabilidad de aprobar el examen mediante la distribucion normal. Veamos como hacerlo encodigo

```python
# parametros de la distribucion normal

# media
media = 100
# desviacion tipica
std = 7.07
```

Vamos a calcular la probabilidad de obtener mas de 110 respuestas correctas:

$$P(X > 125) = 1 - P(X \leq 125)$$

Calculemos la probabilidad acumulada hasta 110

```python
# probabilidad acumulada hasta 125
prob_110 = norm.cdf(110, media, std)

# probabilidad mayor a 125
prob = 1 - prob_110
prob
```

## *Ejemplo 9: Un estudio ha mostrado que, en un cierto barrio, el 60% de los hogares tienen al menos dos televisores. Se elige al azar una muestra de 50 hogares del barrio. Calcule*

### **a.- Cual es la probabilidad de que al menos 20 hogares tengan cuando menos dos televisores?**

Este ejemplo se trata de una distribucion binomial ya que solo existen dos posibles opciones: hogar con dos televisores o diferente de dos televisores. La probabilidad de exito es del 60% mientras que la probabilidad de fracaso es del 40%. La eleccion de los hogares son independientes entre si. Debido a la gran tamano de la muestra, podemos aproximar la distribucion binomial a una normal solo si cumple con las siguiente condiciones:

$$np > 5$$
$$nq > 5$$

Verifiquemos que la D. binomial cumple con los requisitos:

```python
# parametros de la D. binomial

# numero de eventos
n = 50
# probabilidad de exito
p = 0.60
# probabilidad de fracaso
q = 0.40

# evaluacion de los requisitos
print(n*p)
print(n*q)
```

Como se cumplen las condiciones podemos hacer la aproximacion:

$$B(n,p) \sim N(np,\sqrt{npq}) \rightarrow N(30, 3.46)$$

Ahora, calculamos la probabilidad que al menos 20 hogares tengan 2 televisores:

$$P(X > 20) = 1 - P(X \leq 20)$$

Veamos como calcular la probabilidad acumulada con codigo

```python
# parametros de la D. normal

# media
media = 30
# desviacion tipica
std = 3.46

# probabilidad acumulada hasta 20
prob_20 = norm.cdf(20, media, std)

# probablidad superior a 20
prob = 1 - prob_20
prob
```

> La probabilidad de que al menos 20 de los 50 hogares tengan dos televisores es de 99.8% 

### **a.- Cual es la probabilidad de que entre 35 y 40 hogares tengan cuando menos dos televisores?**

Debemos calcular la probabilidad de que entre 35 y 40 hogares tengan cuando menos dos televisores:

$$P(35 \leq X \leq 40) = P(X \leq 40) - P(X \leq 35)$$

```python
# probabilidad acumulada hasta 40
prob_40 = norm.cdf(40, media, std)

# probabilidad acumulada hasta 35
prob_35 = norm.cdf(35, media, std)

# probabilidad entre 35 y 40
prob = prob_40 - prob_35
prob
```

> La probabilidad de que entre 35 y 40 hogares tengan cuando menos dos hogares es de 7.22%

## *Ejemplo 10: El tiempo necesario para que germine una variedad de semillas de una planta esta normalmente distribuido con una media de 15 dias y desviacion tipica de 4 dias. Calcular*

### **a.- Que proporcion de semillas deben germinar antes de 19 dias?**

La distribucion normal, en este caso, viene representada por:

$$N(15, 4)$$

nos piden calcular el porcentaje de semillas que deben germinar antes de 19 dias, es decir, la distribucion acumulada hasta 19.

```python
# parametros de la D. normal

# media
media = 15
# desviacion tipica
std = 4
```

```python
# probabilidad acumulada hasta 19
prob_19 = norm.cdf(19, media, std)
prob_19
```

> El 84.13% de las semillas germinan antes de 19 dias

### **b.- Que proporcion de semillas germinan despues de 23 dias?**

Nos piden calcular el porcentaje de semillas que germinan despues de 23 dias, es decir:

$$P(X > 23) = 1 - P(X \leq 23)$$

Calculemos la probabilidad acumulada hasta 23

```python
# probabilidad acumulada hasta 23
prob_23 = norm.cdf(23, media, std)

# probabilidad superior a 23
prob = 1 - prob_23
prob
```

> El 2.27% de las semillas germinan despues de 23 dias

### **c.- A los cuantos dias deben haber germinado tres cuartas partes de las semillas?**

Las tres cuartas partes de las semillas representan el 75% del total. Debemos calcular la cantidad de dias que contiene el 75% de las semillas germinadas. Para ello utlizamos el metodo *ppf* de *Scipy* como sigue:

```python
# valor que contiene 75% de los datos
val = norm.ppf(0.75, media, std)
val = val.round(0)
val
```

# La Distribucion Normal y su estrecha relacion con el Teorema del Limite Central

Es una de las curiosidades estadisticas mas interesantes que he encontrado y conecta los conceptos de muestra, poblacion y distribucion normal con cualquier otra distribucion de manera elegante. Veamos el enunciado del teorema de limite central para entenderlo mejor:

>  **Establece que, a medida que aumenta el tamaño de la muestra, la distribución de las medias muestrales se aproxima a una distribución normal con una media igual a la media de la población y una desviación estándar $\sigma_{\overline{x}}$ igual a la desviación estándar de la variable original dividida por la raíz cuadrada del tamaño de la muestra. Este hecho es válido para $n$ mayor o igual a 30. La desviacion estandar de las medias muestrales se conoce como error estandar de la media y viene representado por**

$$\sigma_{\overline{x}} = \frac{\sigma}{\sqrt{n}}$$

Vamos a verificar que se cumpla el teorema utilizando un conjunto de datos basado en informacion demograficas de personas que residen en una ciudad.

### **Cargar Datos**

```python
# libreria
import pandas as pd

# obtener datos 
df = pd.read_csv('base.csv')
df.head()
```

```python
print(df.shape)
```

### **Seleccion de la variable y ver su distribucion**

En este caso, vamos a seleccionar la Edad de las personas. Veamos la distribucionque sigue las edades

```python
df.Edad.hist();
```

Vemos que sigue una distribucion normal sesgada a la derecha. Calculemos su media

```python
# media de la poblacion
media_p = df.Edad.mean()
media_p
```

La media de la edad de la poblacion es 44.07

### **Creacion de 1500 muestras de 2000 registros de la variable edad**

```python
# tamaño de muestras
n = 2000
# total de muestras
total_muestras = 1500
# lista para almacenar muestras
muestras_list = []

# generar muestras
for i in range(total_muestras):
    # seleccion de datos aleatorios de Edad
    _ = df.Edad.sample(n)
    # resetear indices de la muestra
    _.index = range(0, len(_))
    # guardar muestra en el df muestras
    muestras_list.append(_.rename('Muestra_' + str(i)))
    
# concatenar todas las muestras en un solo dataframe
muestras = pd.concat(muestras_list, axis=1)

muestras.head()
```

### **Primera parte del teorema**

La primera parte del teorema indica que *a medida que aumenta el tamaño de la muestra, la distribucion de las medias de la muestra se aproxima a una distribucion normal*. Verifiquemos que se cumpla creando el histograma de las medias muestrales.

```python
muestras.mean().hist();
```

Vemos que la distribucion de la media de las muestras tiende a una distribucion normal. Esto confirma que se cumple la primera parte del teorema.

### **Segunda parte del teorema**

La segunda parte del teorema indica que *la media de las medias muestrales es igual a la media de la poblacion*. Calculemos la media de la media de las muestras.

```python
muestras.mean().mean()
```

Vemos que la media de la media de las muestras es de 44.07, similar a la de la poblacion. Al comparar la media tanto de la poblacion (44.07) como de las medias muestrales (44.07 ) confirmamos que se cumple la segunda parte del teorema.

### **Tercera parte del teorema**

La tercera parte del teorema indica que *la desviacion estandar es igual a la desviacion estandar de la variable original dividida por la raiz cuadrada del tamaño de la muestra*. Para verificarlo, calculamos la desviacion estandar de la poblacion y de las muestras.

```python
# desviacion estandar de la poblacion
sigma = df.Edad.std()
sigma
```

```python
# desviacion estandar de las muestras
sigma_x = muestras.mean().std()
sigma_x
```

Verifiquemos que la ecuacion del error estandar de la media se cumpla

$$\sigma_{\overline{x}} = \frac{\sigma}{\sqrt{n}}$$

```python
# error estandar de la media
sigma_x2 = (df.Edad.std())/(np.sqrt(n))
sigma_x2
```

Vemos que el error estandar es igual a la desviacion estandar de las muestras, por tanto, se cumple la tercera parte del teorema.

El teorema permite utilizar formulas de aproximacion normal para calcular distribuciones muestrales con las que realizar inferencias (intervalos de confianza y prueba de hipotesis).

# Diagrama Q-Q

Se utilizan para determinar visualmente lo cerca que esta una muestra de una distribucion especifica, en este caso, de la normal. Su objetivo principal es evaluar si los datos siguen una distribucion teorica especifica, identificar desviaciones de dicha distribucion o comparar las distribuciones de dos conjuntos de datos diferentes.

    2.- Se calculan los cuantiles correspondientes de la distribucion teorica con la que se desea comparar los datos, en este caso para la normal estandar.

    3.- Se grafican los cuantiles de los datos en el eje Y y los cuantiles teoricos en el eje X

    4.- Si los datos siguen una distribucion teorica, los puntos del grafico se alinearan aproximadamente a lo largo de una linea recta diagonal. Desviaciones de esta linea indican diferencias entre la distribucion de los datos y la teorica.
