## 1) Muestreo Aleatorio Simple
Metodo de seleccion de ciertas unidades sacadas de una poblacion de manera que cada una de las muestras tiene la misma probabilidad de ser elegida. Ejemplo, la loteria

```python
# muestra de 8 registros
aleat_8 = econdata.sample(n=8)
aleat_8
```

```python
# creamos otra muestra de 8 elementos para corroborar que funciona correctamente
aleat_8_2 = econdata.sample(n=8)
aleat_8_2
```

```python
# muestra del 25% del total de la poblacion
prop_25 = econdata.sample(frac=.25)
prop_25.head()
```

## 2) Muestreo Sistematico
Metodo de seleccion de ciertas unidades al azar y, a continuacion, se eligen el resto siguiendo intervalos regulares. Se imponen condiciones internas o externas para la seleccion. Ejemplo, dar un premio cada cien personas que hagan una inscripcion hasta llegar a un total de mil inscritos

```python
# creamos una funcion que nos permita elegir los intervalos de eleccion de la muestra
def systematic_sampling(econdata, step):
    indexes = np.arange(0, len(econdata), step=step)
    systematic_sample = econdata.iloc[indexes]
    return systematic_sample
```

```python
# creamos la muestra tomando los registros de 3 en 3
systematic_sample = systematic_sampling(econdata, 3)
systematic_sample.head(10)
```
