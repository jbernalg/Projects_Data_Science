# Distribución Exponencial

Es una distribución de probabilidad contínua utilizada para modelar el tiempo que transcurre en realizar una actividad. Algunas aplicaciones comunes de esta distribución son:

- **Modelado de tiempos entre eventos**: tiempo de espera en colas, vida útil de productos, tiempo de carga y descarga.
- **Análisis de supervivencia**: para modelar el tiempo hasta que ocurre un evento de interés.
- **Confiabilidad**: para modelar el tiempo hasta el fallo de uncomponente o sistema.
- **Riesgo proporcional**: para modelar la tasa de riesgo constante.


## Función de Densidad de Probabilidad

La **Función de Densidad de Probabilidad** viene dada por:

$$f(x) = \left\{\begin{matrix}
\lambda e^{-\lambda x} & si & x\geq 0\\ 0  
 & si & x < 0 
\end{matrix}\right.$$

donde $\lambda > 0$ es la tasa de eventos por unidad de tiempo. Esta distribución se aplica solo para valores positivos de $x$ y situaciones donde los valores más pequeños de $x$ sean más probable que valores mayores. La grafica de esta distribución viene dada por:



La probabilidad, como hemos visto en otras distribuciones, se obtiene mediante el cálculo del área bajo la curva. Calculemos toda el área bajo esta curva mediante integrales.

$$A =\int_{-\infty}^{\infty}f(x)dx = \int_{-\infty}^{\infty}\lambda e^{-\lambda x}dx$$

$$A =\int_{-\infty}^{0}0 dx + \int_{0}^{\infty}\lambda e^{-\lambda x}dx$$

$$A =\left [\frac{-\lambda}{\lambda} e^{-\lambda x} \right ]$$

Ahora, evaluamos el resultado de la integral en $0$ e $\infty$

$$A = \lim_{x \rightarrow \infty}(-e^{-\lambda x}) - (-e^{0})=0-(-1) = 1$$

El área bajo la curva es de 1, lo cual garantiza que podemos utilizar esta distribución para calcular probabilidades.

## Función de Distribución de Probabilidad Acumulada

La **Función de Distribución Probabilidad Acumulada de la Exponencial** es la siguiente:

$$f(x,\lambda) = \left\{\begin{matrix}
1 - e^{-\lambda x} & si & x\geq 0\\ 0  
 & si & x < 0 
\end{matrix}\right.$$

## Cálculo de Probabilidades

- Para calcular la probabilidad de que $x$ sea menor o igual a un valor $x_{0}$ dado, se utiliza
    $$P(x\leq x_{0}) = 1 - e^{-\lambda x_{0}}$$
    
    

- Para calcular la probabilidad de que $x$ sea mayor o igual a un valor $x_{0}$ dado, se utiliza

    $$P(x\geq x_{0}) = e^{-\lambda x_{0}}$$

    

## Parámetros de la Distribución Exponencial

- La media $(\mu)$ de esta ditribución viene dada por:
$$\mu = \frac{1} {\lambda}$$

- La varianza $(\sigma^2)$ viene dada por:
$$\sigma^2 = \frac{1} {\lambda^2}$$

Veamos algunos ejemplos para entender por completo la Distribución Exponencial.

## *Ejemplo 1: El tiempo que tarda para cargar un camión obedece a una distribución exponencial. La media del tiempo de espera para cargarlo es de 15 minutos*

### a.- Cuál es la probabilidad de que la carga de un camión dure 6 minutos o menos?

La media de tiempo de espera para cargar un camión es de 15 minutos. Nos piden calcular la probabilidad de que el tiempo de espera sea menor o igual a 6 minutos, cuya área viene representada por:



Vamos a utilizar la fórmula:

$$P(x\leq x_{0}) = 1 - e^{-\lambda x_{0}}$$

Donde $x_{0}$ es 6 y $\lambda$ es $1/15$. Hagamos el cálculo matemático para obtener el valor de probabilidad.

```python
# libreria para importar la exponencial
from math import exp

P_6 = 1 - exp(-(6)*(1/15))
P_6
```

> La probabilidad de que un camión sea cargado en un tiempo menor o igual a 6 minutos es de 32.97%

### b.- Cuál es la probabilidad de que la carga de un camión dure 18 minutos o menos?

El área que nos piden calcular es la siguiente:



La probabilidad a calcular, mediante la fórmula, es $P(x\leq 18)$. Hagamos el cálculo matemático.

```python
P_18 = 1 - exp(-(18)*(1/15))
P_18
```

También podemos utilizar función de distribucion acumulada de la exponencial dada por la librería *Scipy* para realizar el cálculo con más facilidad.

```python
# parametros de la distribucion

# media
media = 15

# lambda
lamb = 1/media
```

```python
# importar libreria 
from scipy.stats import expon

# calcular la probabilidad acumulada hasta 18
P_18 = expon.cdf(18, scale=1/lamb)
P_18
```

> La probabilidad de que la carga del camión dure menos de 18 minutos es de 69.8%

### c.- Cuál es la probabilidad de que la carga de un camión dure de 4 a 20 minutos?

El área que nos piden calcular es la encerrada entre 6 y 18 tal como se muestra



La probabilidad que nos piden es:

$$P(4 \leq x \leq 20) = P(x \leq 20) - P(x \leq 4)$$

Vamos a calcular la probabilidad acumuluda hasta 20 y luego hasta 4 para finalmente restar ambas probabilidades, usando *Scipy*.


```python
# probabilidad acumulada hasta 20
P_20 = expon.cdf(20, scale=1/lamb)

# probabilidad acumulada hasta 4
P_4 = expon.cdf(4, scale=1/lamb)

# probabilidad entre 4 y 20
prob = P_20 - P_4
prob
```

> La probabilidad de que un camión se cargue entre 4 y 20 minutos es de 50.23%

### b.- Cuál es el tiempo de revisión de un motor superado por el 10% de los tiempos de revisión?

Lo que nos piden es encontrar aquel valor $x_{0}$ de la variable que acumula a su derecha una probabilidad de 0.1.



Esto equivale a decir que ese valor de variable acumulada a su izquierda una probabilidad de 0.9.

$$P(x \leq x_{0}) = 0.9$$

Podemos utilizar la siguiente fórmula, despejando $x_{0}$

$$P(x\leq x_{0}) = 1 - e^{-\lambda x_{0}}$$

Veamos cómo resolverlo algebráicamente:

$$0.9 = 1 - e^{-\lambda x_{0}}$$

$$e^{-\lambda x_{0}} = 0.1$$

$$-\lambda x_{0} = ln(0.1)$$

$$x_{0} = \frac{-ln(0.1)} {\lambda}$$

Ahora procedemos a sustituir el valor de $\lambda$ y resolver el logaritmo

```python
# libreria para calcular logaritmo natural
from math import log

X_0 = (-log(0.1))/lamb
X_0
```

> El tiempo de revisión que supera el 10% de los tiempos de revisión es 51 minutos.

También podemos utilizar el método *ppf* de *Scipy* para realizar este cálculo de manera más eficiente. Veamos como implementarlo.

```python
# valor que contiene el 90% de probabilidad
X_0 = expon.ppf(0.9, scale=1/lamb)
X_0
```

> Obtenemos el mismo valor que el cálculo anterior ahorrando operaciones matemáticas y líneas de código

### c.- El costo de revisión es de 200 unidades monetarias fijas al que se le suma 10 unidades monetarias por el tiempo que dure la revisión. Encontrar la media y la varianza del costo

Representamos a $C$ como el costo de reparación y $X$ el tiempo de reparación. Entonces el costo viene dado por:

$$C = 200 + 10X$$

Para obtener la media del costo, usamos la linealidad de la esperanza matemática:

$$E[C] = E[200 + 10X] = 200 + 10E[X]$$

Dado que $E[X]$ es la media de la distribución exponencial, que es de 22 minutos obtenemos:

$$E[C] = 200 + 10\cdot22 = 200 + 220 = 420$$

La media del costo de revisión es de 420 unidades monetarias.

Para obtener la varianza del costo, usamos la linealidad de la varianza matemática:

$$V[C] = V[200 + 10X] = 10^2V[X] = 100 \cdot V[X]$$

$V[X]$ representa la varianza de la distribución la cual, se calcula con: 

$$\sigma^2 = \frac{1}{\lambda^2}$$

Dado que $\lambda$ es $1/22$, entonces

$$V[X] = 22^2 = 484$$

La varianza del costo es:

$$V[C] = 100 \cdot 484 = 48400$$

## *Ejemplo 4: El tiempo de vida de una lámpara especial sigue una distribución exponencial con media de 100 horas.*

### a.- Cuál es la probabilidad de que una lámpara dure por lo menor 30 horas?

Sabemos que la media $\mu$ es de 100 horas, por tanto

$$\lambda = \frac{1}{\mu} = \frac{1}{100}$$

el área de la distribución exponencial que nos piden calcular es la siguiente:



La probabilidad a hallar es la siguiente:

$$P(X > 30) = 1 - P(X \leq 30)$$

Calculamos la probabilidad acumulada hasta 30

```python
# parametros de la distribucion

# media
media = 100
# lambda
lamb = 1/100
```

```python
# probabilidad acumulada hasta 30
p_30 = expon.cdf(30, scale=1/lamb)

# calcular la probabilidad de mas de 30
prob = 1 - p_30
prob
```

> La probabilidad de que una lámpara dure al menos 30 horas es de 74.08%

### b.- Si una lámpara ya lleva 50 horas de uso. Cuál es la probabilidad de que dure más de 80 horas?

La probabilidad que piden es que una lámpara dure más de 80 horas cuando lleva 50 horas de uso, es decir, es una probabilidad condicionada

$$P(X > 80 | X > 50) = \frac{P(X > 80)}{P(X > 50)}$$

Cada probabilidad se puede expresar como:

$$P(X > 80) = 1 - P(X \leq 80)$$

$$P(X > 50) = 1 - P(X \leq 50)$$

Calculamos ambas probabilidades

```python
# probabilidad acumulada hasta 80
p_80 = expon.cdf(80, scale=1/lamb)
# probabilidad para mas de 80
prob_80 = 1 - p_80
prob_80
```

```python
# probabilidad acumulada hasta 50
p_50 = expon.cdf(50 , scale=1/lamb)
# probabildiad de mas de 50
prob_50 = 1 - p_50
prob_50
```

Ahora, calculamos la probabilidad condicionada

```python
prob_cond = prob_80/prob_50
prob_cond
```

> La probabilidad que una lámpara dure más de 80 horas al haber durado ya 50 horas es de 74.08%. Vemos que este resultado es similar al anterior y se debe a la propiedad de falta de memoria que caracteriza a la distribución exponencial: **La historia pasada no afecta la probabilidad futura de supervivencia en términos de tiempo** 

$$P(X > s + t | X > s) = P(X > t)$$

*La probabilidad de que la variable $X$ sobreviva más de $s + t$ unidades de tiempo, dado que ya haya sobrevivido más de $s$ unidades de tiempo, es igual a la probabilidad original de que $X$ sobreviva más de $t$ unidades de tiempo*

Si aplicamos esta propiedad al ejemplo, la probabilidad a buscar sería:

$$P(X > 80 | X > 50) = P(X > 30)$$

Tal probabilidad $P(x > 30)$ es la misma calculada en el inciso **a**.

### c.- Se seleccionan 5 lámparas. Cuál es el número esperado de lámparas que duran por lo menos 30 horas de las 5 consideradas?

Podemos considerar este problema como uno de distribución binomial $B(n,p)$ con dos posibles opciones: duran o no duran más de 30 horas. Consideramos el número de eventos $n$ como la cantidad de lámparas y la probabilidad como aquella en la que las lámparas duran más de 30 horas:

$$B(5, 0.7408)$$

Para calcular el número de lámparas que podrían durar más de 30 horas, operamos como sigue
$$n \cdot p$$

Entonces, el promedio de lámparas que duran más de 30 horas es:

> Aproximadamente 4 de las 5 lámparas duran más de 30 horas

## *Ejemplo 6: El tiempo que los cónyuges dedican a la compra de tarjetas de aniversario se puede modelar mediante una distribución exponencial con un promedio de tiempo igual a 8 minutos. Escriba la distribución y haga un gráfico de la distribución y la funcion de densidad.*

La media de la distribución es $\mu = 8$. La función de distribución en este caso, viene dado por:

$$f(x) = \frac{1}{\mu}e^{-\frac{1}{\mu}x} = \frac{1}{8}e^{-\frac{1}{8}x}$$

Ahora graficamos la distribución exponencial del ejercicio


```python
# parametros de la distribucion

# media
media = 8
#lambda
lamb = 1/media

# librerias a utlizar
import numpy as np
import matplotlib.pyplot as plt

# crear 1000 puntos entre 0 y 40 que sigan una distribucion exponencial
x = np.linspace(0, 40, 1000)
y = expon.pdf(x, scale=media)

# crear grafica
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribucion Exponencial (media = {media})', color='blue')

# Añadir una línea vertical en la media
plt.axvline(x=media, color='red', linestyle='--', label='Media (8 horas)')

plt.title('Distribucion Exponencial', fontsize = 18)
plt.xlabel('x', fontsize=14)
plt.ylabel('Densidad', fontsize = 14)
plt.legend();
```

Ahora vamos a graficar la función de densidad de la distribución exponencial

```python
# parametros de la distribucion

# media
media = 8
#lambda
lamb = 1/media

# librerias a utlizar
import numpy as np
import matplotlib.pyplot as plt

# crear 1000 puntos entre 0 y 40 que sigan una distribucion exponencial
x = np.linspace(0, 40, 1000)
y = expon.cdf(x, scale=media)

# crear grafica
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Exponencial CDF (media = {media})', color='blue')

# Añadir una línea vertical en la media
plt.axvline(x=media, color='red', linestyle='--', label='Media (8 horas)')

plt.title('Distribucion Exponencial CDF', fontsize = 18)
plt.xlabel('x', fontsize=14)
plt.ylabel('Probabilidad Acumulada', fontsize = 14)
plt.legend();
```
