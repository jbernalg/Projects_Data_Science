# **Introduccin a los algoritmos y herramientas de trabajo**

### Qu es un algoritmo?

* Un Algoritmo es una secuencia de operaciones detalladas y no ambiguas, finitas, que al ejecutarse  paso a paso, conducen a la solucin de un problema.


* Algoritmo es un conjunto de instrucciones que especifican la secuencia de operaciones a realizar, en orden, para resolver un sistema especfico o clase de problema.


* Un Algoritmo es la aplicacin de pasos lgicos, secuenciales y metdicamente aplicados para dar solucin a un problema en cuestin. En otras palabras un algoritmo es una frmula para resolver problemas.


Un algoritmo se puede ver como un bloque de proceso que transforma <b>datos de entrada</b> en <b>informacin de salida</b>. Es decir, se <b>ingresa uno o varios datos</b> y el algoritmo <b>muestra el o los resultados</b> de lo hecho con esos datos.
