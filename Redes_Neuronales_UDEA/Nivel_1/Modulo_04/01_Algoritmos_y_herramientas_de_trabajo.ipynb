# **Introducción a los algoritmos y herramientas de trabajo**

### ¿Qué es un algoritmo?

* Un Algoritmo es una secuencia de operaciones detalladas y no ambiguas, finitas, que al ejecutarse  paso a paso, conducen a la solución de un problema.


* Algoritmo es un conjunto de instrucciones que especifican la secuencia de operaciones a realizar, en orden, para resolver un sistema específico o clase de problema.


* Un Algoritmo es la aplicación de pasos lógicos, secuenciales y metódicamente aplicados para dar solución a un problema en cuestión. En otras palabras un algoritmo es una fórmula para resolver problemas.


Un algoritmo se puede ver como un bloque de proceso que transforma <b>datos de entrada</b> en <b>información de salida</b>. Es decir, se <b>ingresa uno o varios datos</b> y el algoritmo <b>muestra el o los resultados</b> de lo hecho con esos datos.

<h3>Por otra parte...</h3>

Los algoritmos son independientes tanto del lenguaje de programación en que se expresan, como de la arquitectura de la computadora que los ejecuta. Es decir que, <b><i>un lenguaje de programación es tan sólo un medio para expresar un algoritmo y una computadora es sólo un procesador para ejecutarlo</i></b>.
