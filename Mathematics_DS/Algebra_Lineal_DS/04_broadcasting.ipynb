# Suma de matrices y vectores (Broadcasting)

El Broadcasting consiste en extender la matriz o vector de menor tamao para que se ajuste a la de mayor tamao.



```python
# importamos librerias
import numpy as np
```

Definimos algunos elementos

```python
scalar = 5.679
vector = np.array([3,4,5])
matriz = np.array([[1,2],[3,4],[5,6]])
```

Si sumamos "vector" con "matriz" nos arroja error debido a que las dimensiones no coinciden.

```python
A = vector + matriz
print(A)
```

Ahora, si sumamos "matriz" transpuesta con "vector" la dimension de la columna de la matriz coincide con la dimension columna del vector, por tanto, realiza la operacion.

```python
A = matriz.T + vector
print(A)
```

El broadcasting tambien se presenta cuando sumamos a una matriz un escalar.

```python
C = matriz + scalar
print(C)
```
