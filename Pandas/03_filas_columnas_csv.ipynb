## Seleccion de filas y columnas en un archivo CSV

```python
# librerias
import pandas as pd
```

```python
# cargar los datos
datos = pd.read_csv('./data/ATP.csv')
```

```python
# informacion resumida sobre el dataframe
datos.info()
```

```python
datos.head()
```

#### Seleccionar N filas

```python
# selecciona las primeros 5 lineas 
datos.iloc[0:5]
```

```python
# Seleccionar filas especificas a partir de una lista
datos.iloc[[0,3,6,9,12],]
```

```python
# seleccionar filas y columnas a partir de listas

# seleccionar filas 10, 20 30 y 40 y columnas date y surface
datos.iloc[[10,20,30,40], [3,6]]
```

### Seleccionar N columnas

```python
datos.iloc[:,0:2]
```

```python
#Seleccionar filas y columnas
datos.iloc[[0,3,6,9], [0,1,2]]
```

```python
#Seleccionar rango de filas y columnas
datos.iloc[0:5,5:8]
```
