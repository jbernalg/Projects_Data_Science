## Agregar una columna a un DataFrame

```python
#librerias
import pandas as pd
```

```python
# cargar datos
datos = pd.read_csv('./data/energy.csv', header=0) # header 0 indica que la primera linea se toma como nombre de las columnas
df = pd.DataFrame(datos)
df.head()
```

### Convertir a Kelvin los grados de T1

```python
Tk = df['T1'] + 273.15

# asignar los valores de Tk en una nueva columna
df = df.assign(kelvin=Tk.values)
```

```python
df.head()
```

### Guardar resultados en archivo csv

```python
df.to_csv('Energy2.csv', sep='\t')
```
