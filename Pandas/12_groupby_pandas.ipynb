## Agrupar datos con Groupby

```python
# librerias
import pandas as pd
```

```python
# cargar datos
datos = pd.read_csv('./data/movies2.csv')

# lo volvemos en un dataframe
df = pd.DataFrame(datos)
df
```

```python
df.columns
```

### 1. Cual es la maxima duracion de una pelicula?

```python
df['duration'].max()
```

```python
# convertir la duracion en horas
minutos = df['duration'].max()
horas = minutos/60
horas
```

### 2. Cual es esa pelicula que tiene la duracion maxima?

```python
# indice de la pelicula con maxima duracion
df['duration'].idxmax()
```

```python
# obtener el nombre de la pelicula
#_get_value(index, columns) utilizado para obtener un valor dado el indice y la columna de donde
# se quiere obtenr dicho valor
df._get_value(df['duration'].idxmax(), 'movie_title')
```

### 3. Uso de Groupby. Comparar el numero de likes por genero de pelicula

```python
#       columna como indice     columna como valor    operacion sobre los valores
df1 = df.groupby(['genres'])[['movie_facebook_likes']].sum()
df1
```

### 4. Genero mas gustado en Facebook

```python
likes = df.groupby(['genres'])[['movie_facebook_likes']].sum()
likes['movie_facebook_likes'].idxmax()
```

### 5. Cantidad de likes recibidos por el genero mas gustado

```python
likes['movie_facebook_likes'].max()
```

### 6. cual es el promedio de presupuesto gastado por director?

```python
# suma del presupuesto gastado por director
df2 = df.groupby(['director_name'])[['budget']].sum()
df2
```

```python
# promedio gastado por director
df2 = df.groupby(['director_name'])[['budget']].mean()
df2
```

### 7. Cual es el genero que gasta mas dinero?

```python
dinero = df.groupby(['genres'])[['budget']].sum()
dinero
```

```python
# mostrar el genero que mas gasta presupuesto
dinero['budget'].idxmax()
```

```python
# mostrar el monto del presupueso maximo
dinero['budget'].max()
```

### 8. Cual es el actor principal que atrae mas ganancias?

```python
actor = df.groupby(['actor_1_name'])[['gross']].sum()
actor
```

```python
# nombre del actor
actor['gross'].idxmax()
```

```python
# entrada de dinero
actor['gross'].max()
```
