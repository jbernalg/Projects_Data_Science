## Ordenar valores por columna

```python
# liberias
import pandas as pd
import numpy as np
```

```python
#cargar datos
df = pd.read_csv('./data/DatosYT.csv')
```

```python
df.head()
```

```python
# tipos de datos
df.dtypes
```

### Ordenar valores segun una columna

```python
df.sort_values(by=['id'], ascending=True) # ordena valores de menor a mayor 
```

```python
df.sort_values(by=['xs'], ascending=False) # ordena valores de mayor a menor
```

### Ordenar todas las columnas

```python
# ordena las columnas de manera independiente a las demas
df_1 = pd.DataFrame(np.sort(df.values, axis=0),
                    index=df.index, 
                    columns=df.columns)
df_1
```
