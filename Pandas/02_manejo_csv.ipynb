## Estadistica Basica para un archivo CSV

```python
# librerias
import pandas as pd
import numpy as np
```

```python
# Cargar los datos del csv
datos = pd.read_csv('./data/ATP.csv')
datos.head()
```

```python
# Informacion sobre los datos
datos.info()
```

```python
# mostrar solo n registros 
datos.head(5)
```

### Estadistica de los datos

```python
datos.describe()
# aplica para 42 variables de las 54 debido a que son de tipo numerico
```

### Sustituir valores faltantes

```python
df = pd.DataFrame(datos)
```

```python
df = df.replace(np.nan, '0')
df = df.replace('N/A', '0')
df = df.replace('NR', '0')
df.info()
```

### Estadisiticas de los datos sin NaN

```python
df.describe()
# solo toma dos variables porque son de tipo numerico
```

### Nombre de las variables

```python
# Con list. Muestra los nombres en formato de lista: 1 nombre por fila
list(df)
```

```python
# con el metodo columns. Muestra los dato en formato de lista continua: un dato seguido del otro
df.columns
```

### Convertir variables tipo objects a numericas

```python
df['Wsets'] = df.Wsets.astype(int)
df['WRank'] = df.WRank.astype(int)
df.describe()
```

### Eliminar registros que contengan valores faltantes

```python
# Tipo de datos de la columna con valores a eliminar
df['LBL'].dtypes
```

```python
# eliminar todos lo registros que tenga 0 como valor en la columna LBL
df.drop(df[df['LBL'] == '0'].index, inplace=True)
df.head()
```
