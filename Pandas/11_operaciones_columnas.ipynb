## Operaciones con condicion en Columnas

```python
#librerias
import pandas as pd
import numpy as np
```

```python
# cargar datos
datos = pd.read_csv('./data/minutos2.csv')
df = pd.DataFrame(datos)
df
```

### Crear una nueva columna con la diferencia entre valores de otra columna 

```python

#inciamos el valor alto con la fila 0 de la columna B
alto = df.loc[0, 'B']

#iterar sobre todos los valores de B
for i in range(1, len(df)):
    # si el valor de B de la fila i es mayor al valor anterior y si el 
    # valor de b de la fila i es mayor a alto entonces asignar a la nueva columna DB
    # la diferencia entre ese valor y alto 
    if (df.loc[i, 'B'] > df.loc[i - 1, 'B']) and (df.loc[i, 'B'] > alto):
        df.loc[i, 'dB'] = df.loc[i, 'B'] - alto
        alto = df.loc[i, 'B']
    else:
        df.loc[i, 'dB'] == 0
```

```python
df
```

### Sustituir los valores NaN por cero

```python
df = df.replace(np.nan, 0)
df
```

### Sumar valores de las diferencias

```python
total = df['dB'].sum()
total
```

### Crear nueva columna con las diferencias entre los valores de la columna B

```python
df['PB'] = df['B'].diff(1)
df
```

### Sumar solo los valores positivos de la columna PB

```python
suma_PB = df.loc[df['PB'] > 0, 'PB'].sum()
suma_PB
```
