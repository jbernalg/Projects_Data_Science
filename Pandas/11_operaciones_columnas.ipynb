## Operaciones con condicion en Columnas

```python
#librerias
import pandas as pd
```

```python
# cargar datos
datos = pd.read_csv('./data/minutos2.csv')
df = pd.DataFrame(datos)
df
```

### Crear una nueva columna con la diferencia entre valores de otra columna 

```python

#localizar los valores altos de la columna B
alto = df.loc[0, 'B']

#iterar sobre todos los valores de B
for i in range(1, len(df)):
    if (df.loc[i, 'B'] > df.loc[i - 1, 'B']) and (df.loc[i, 'B'] > alto):
        df.loc[i, 'dB'] = df.loc[i, 'B'] - alto
        alto = df.loc[i, 'B']
    else:
        df.loc[i, 'dB'] == 0
```

```python
df
```

### Sumar valores de las diferencias

```python
total = df['dB'].sum()
total
```

### Crear nueva columna con las diferencias entre los valores de la columna B

```python
df['PB'] = df['B'].diff(1)
df
```

### Sumar solo los valores positivos de la columna PB

```python
suma_PB = df.loc[df['PB'] > 0, 'PB'].sum()
suma_PB
```
