## Graficas de Pie y de Barra

```python
# librerias
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# cargar datos
df = pd.read_csv('./data/VGMod.csv')
df.head()
```

```python
# lista de colores a utilizar
colors  = ("dodgerblue","salmon", "palevioletred", 
           "steelblue", "seagreen", "plum", 
           "blue", "indigo", "beige", "yellow")
```

### Grafica de Pie

```python
# iterar sobre el df para obtener la grafica de cada columna
i = 0
for col in df:
    sizes = df[col].value_counts()
    pie = df[col].value_counts().plot(kind='pie', # tipo de grafica
            colors=colors,                        # colores a utilizar
            shadow=True,                          # sombra de la grafica
            autopct='%1.1f%%',                    # formato de porcentaje
            startangle=30,                        # angulo de la grafica
            radius=.5 ,                           # distancia del grafico y la sombra
            center = (0.5, 0.5),                  # centro de la grafica
            textprops={'fontsize':12},            # tamaño del texto de las etiquetas
            frame=False,                          # recuadro de etiquetas
            labels=None,                          # elimina las etiquetas de la grafica
            pctdistance = .65                     # distancia del pct al centro del pie
            )
    labels = sizes.index.unique()                 # muestra la etiqueta de valores unicos
    plt.gca().axis('equal')                       # distribucion equitativa de los ejes
    plt.title(df.columns[i], weight='bold', size=14) # titulo de cada grafica
    plt.subplots_adjust(
        left=0.0,
        bottom=0.1,
        right=0.85
    )
    # plt.savefig(
    #     './data/'+str(df.columns[i])+'png',       # ruta y nombre de la imagen
    #     dpi=100,                                  # definicion de la imagen
    #     bbox_inches='tight'                       # ajuste del titulo de la grafica
    # )
    pie.set_ylabel('')                              # elimina el titulo del eje y
    plt.legend(labels, bbox_to_anchor=(0.5, -.2))   # leyenda con los colores de la grafica
    i += 1
    plt.show()
```

### Grafica de Barras

```python
# iterar sobre el df para obtener la grafica de cada columna
i = 0
for col in df:
    sizes = df[col].value_counts()
    graph = df[col].value_counts().plot(kind='bar')

    labels = sizes.index.unique()                 # muestra la etiqueta de valores unicos
    plt.title(df.columns[i], weight='bold', size=14) # titulo de cada grafica
    plt.subplots_adjust(
        left=0.0,
        bottom=0.1,
        right=0.85
    )
    # plt.savefig(
    #     './data/'+str(df.columns[i])+'png',       # ruta y nombre de la imagen
    #     dpi=100,                                  # definicion de la imagen
    #     bbox_inches='tight'                       # ajuste del titulo de la grafica
    # )
    graph.set_ylabel('')                              # elimina el titulo del eje y
    i += 1
    plt.show()
```
