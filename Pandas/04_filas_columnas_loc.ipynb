## Seleccion de filas y columnas por el nombre

```python
#librerias
import pandas as pd
```

```python
# cargar los datos
datos = pd.read_csv('./data/ATP.csv')
```

```python
# mostrar primeras lineas del dataframe
datos.head()
```

### Fijar una columna como indice del dataframe

```python
datos.set_index('Location', inplace=True)
datos.loc['Melbourne']
```

```python
# traer valores de filas con valores especificos de una columna

# mostrar los valores de Surface para Atlanta
datos.loc['Atlanta', 'Surface']
```

```python
# Mostrar Winner para Location Adelaide
datos.loc['Adelaide', 'Winner']
```

### Seleccion de filas y columnas especificas

```python
#           Valores de indice        Valores de columna
datos.loc[['Atlanta', 'Melbourne'], ['Series', 'Court']]
```

```python
datos.loc[['Chennai','Adelaide'], ['Winner','Round']]
```

### Seleccion de filas y columnas por rangos

```python
#seleccionar todas las filas de atlanta y melbourne y las columnas que van de Series a Round
datos.loc[['Atlanta', 'Melbourne'], 'Series':'Round'] 
```

```python
datos.loc[['Chennai','Adelaide'], 'Date':'Loser']
```

### Seleccion de filas segun valor de una columna

```python
#seleccionar filas que sea del Gran Slam
datos.loc[datos['Series'].str.endswith('Slam')]
```

```python
# seleccionar registros que terminen con Round en Round
datos.loc[datos['Round'].str.endswith('Round')]
```

```python
# mostrar registros que tenga como valor Indoor en la columna Court
datos.loc[datos['Court'].str.endswith('Indoor')]
```
